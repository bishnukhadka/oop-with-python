## Part 1 — Types of Errors

### The Three Types of Python Errors

| Type | When It Occurs | Python Detects It? | Can You Handle It? |
|---|---|---|---|
| **Syntax Error** | Before program runs | Yes — won't even start | No — must fix the code |
| **Runtime Exception** | During execution | Yes — program crashes | ✅ Yes — use `try/except` |
| **Logical Error** | During execution | No — runs but wrong result | No — must test carefully |

### 1.1 Syntax Errors

- A **syntax error** means Python cannot parse your code.
- It refuses to run at all.
- It is like writing a sentence with completely broken grammar.

**Analogy:** Texting *"Can you pick from the car me up?"* — completely unreadable.

````python
# ── 1.1 SYNTAX ERRORS ────────────────────────────────────────────────────────
# A syntax error prevents Python from running your program at all.
# Python's compile() lets us demonstrate syntax errors safely in a notebook.

print("=== Syntax Error Demonstration ===\n")

x=1
if x > 10 # Here is an error
    print("x is greater than 10")
````

### 1.2 Runtime Errors (Exceptions)

- A **runtime error** occurs **while** the program is running. 
- The syntax is valid but something unexpected happens. 
- You can **catch and handle** these.

**Analogy:** Clear directions: *"Turn left at the traffic light."* Valid, but what if
the light has been removed? You only discover the problem when you make the journey.

````python
# ── 1.2 RUNTIME ERRORS (EXCEPTIONS) ─────────────────────────────────────────
# The syntax is valid, but something unexpected happens during execution.

# ZeroDivisionError
print("--- ZeroDivisionError ---")
try:
    total_marks = 450
    num_subjects = 0
    average = total_marks / num_subjects   # Cannot divide by zero!
    
except ZeroDivisionError as e:
    print(f"  Caught : {e}")
    print("  Fix    : Check num_subjects > 0 before dividing\n")
````

## Part 2 — Exception Handling: try / except / else / finally

### How It Works

```
try:
    ← risky code — might raise an exception
except ExceptionType:
    ← what to do when that exception occurs
else:
    ← runs ONLY if no exception occurred (success path)
finally:
    ← ALWAYS runs — cleanup code goes here
```

### Flow Diagram

```
try block runs
  ├─ No exception? → else block → finally block → continue
  └─ Exception?    → except block → finally block → continue
```

> **Key Principle:** The program does **NOT crash**.
> After the except/finally block, execution continues normally.

````python
# ── 2.1 BASIC try/except ─────────────────────────────────────────────────────
# Without exception handling → program crashes on bad input.
# WITH exception handling → program survives and continues.

try:
    num = int(input("Enter a number: "))
    print(f'Inputted number = {num}')
    result = 10 / num

except ZeroDivisionError:
    print("Exect block ran: You cannot divide by zero.")

except ValueError:
    print("Except block ran: Please enter a valid number.")

else:
    print("Else block ran (No Exception Occured): Result:", result)
finally:
    print("Finally block ran: Program finished.")
````

`type(e).__name__`: returns the class name as a string

````python
try:
    num = int(input("Enter a number: "))
    print(f'Inputted number = {num}')
    result = 10 / num

except Exception as e:
    print("An error occurred!")
    print("Error message:", e)
    print("Error type:", type(e).__name__)

# type(e) -> <class `ZeroDivisionError`>
# type(e).__name__ -> ZeroDivisionError
````

If a finally clause is present, the finally clause will execute as the last task before the try statement completes. The finally clause runs whether or not the try statement produces an exception. The following points discuss more complex cases when an exception occurs:
- If an exception occurs during execution of the try clause, the exception may be handled by an except clause. If the exception is not handled by an except clause, the exception is re-raised after the finally clause has been executed.
- An exception could occur during execution of an except or else clause. Again, the exception is re-raised after the finally clause has been executed.
- If the finally clause executes a break, continue or return statement, exceptions are not re-raised. This can be confusing and is therefore discouraged. From version 3.14 the compiler emits a SyntaxWarning for it (see PEP 765).
- If the try statement reaches a break, continue or return statement, the finally clause will execute just prior to the break, continue or return statement’s execution.
- If a finally clause includes a return statement, the returned value will be the one from the finally clause’s return statement, not the value from the try clause’s return statement. This can be confusing and is therefore discouraged. From version 3.14 the compiler emits a SyntaxWarning for it (see PEP 765).

Reference: [docs.python.org](https://docs.python.org/3/tutorial/errors.html#defining-clean-up-actions)

### ⚠️ Common Mistakes in Exception Handling

| Mistake | Why It's Wrong | The Fix |
|---|---|---|
| `except:` (bare) | Catches **everything** — even `Ctrl+C` | Always name the exception type |
| `except Exception:` first | Shadows specific handlers below — they're never reached | Put broad handlers **last** |
| `except ...: pass` only | Silently swallows the error — impossible to debug | At minimum, `print(error)` |
| Too much code in `try` | Can't tell which line actually failed | Keep `try` blocks **short and focused** |
| No `finally` for cleanup | Resources stay open or locked | Use `finally` or the `with` statement |

```python
# ❌ NEVER do this — catches KeyboardInterrupt and SystemExit too
try:
    risky_operation()
except:
    pass

# ✅ Always name the exception
try:
    risky_operation()
except ValueError:
    print("Handle ValueError...")
except TypeError:
    print("Handle TypeError...")
```

#### SEPARATE except CLAUSES 

Different exceptions → different, tailored responses.

````python


print("=== Separate except Clauses ===\n")

def look_up_score(grade_book, student_key):
    """Look up a student score — handle two separate error types."""
    try:
        student_id = int(student_key)      # Could raise ValueError
        score      = grade_book[student_id]  # Could raise KeyError
        print(f"\n---\n  Student {student_id}: {score}")
        return score
    except ValueError:
        print(f"\n---\n  Error: '{student_key}' is not a valid ID (must be a number).")
        return None
    except KeyError:
        print(f"\n---\n  Error: Student ID {student_key} does not exist in the grade book.")
        return None

grade_book = {101: 85, 102: 72, 103: 91}

print("--- Grade Book Lookups ---")
look_up_score(grade_book, "101")    # Valid   → 85
look_up_score(grade_book, "999")    # KeyError → not found
look_up_score(grade_book, "abc")    # ValueError → not a number
look_up_score(grade_book, "102")    # Valid   → 72
````

#### TUPLE GROUPING
When two different exceptions deserve the SAME response, group them in a tuple.

````python


print("=== Tuple Grouping ===\n")

def add_to_cart(cart, items, index_str):
    """Add item[index_str] to the cart."""
    try:
        index = int(index_str)     # ValueError if not a number
        item  = items[index]       # IndexError if out of range
        cart.append(item)
        print(f"  Added '{item}' to cart.")
    except (ValueError, IndexError):
        # Same helpful message for both errors
        print(f"  Invalid selection '{index_str}'. "
              f"Enter a number 0–{len(items) - 1}.")
    except Exception as e:
        # Unexpected errors — always put this LAST
        print(f"  Unexpected error [{type(e).__name__}]: {e}")

available = ["Apples", "Bread", "Milk", "Cheese", "Eggs"]
cart = []

print(f"Available: {available}\n")
add_to_cart(cart, available, "1")     # Valid   → Bread
add_to_cart(cart, available, "abc")   # ValueError
add_to_cart(cart, available, "99")    # IndexError
add_to_cart(cart, available, "4")     # Valid   → Eggs
add_to_cart(cart, available, "-1")    # Valid Python (last element) — design choice!

print(f"\nFinal cart: {cart}")
````

#### 3.3 EXCEPTION ORDER MATTERS 

- SPECIFIC exceptions must come BEFORE broad ones.
- Python checks except clauses in order — first match wins.

````python
print("=== Exception Order Matters ===\n")

# ❌ WRONG ORDER — ZeroDivisionError is a subclass of ArithmeticError,
# which is a subclass of Exception. If Exception is first, it always matches.
print("--- Wrong order (dead code) ---")
def bad_order(a, b):
    try:
        return a / b
    except Exception as e:
        print(f"  Generic: {e}")
    except ZeroDivisionError:    # NEVER reached — Python stops at Exception
        print("  ZeroDivisionError (you will never see this line)")

bad_order(10, 0)   # Shows "Generic: division by zero" — not ideal

# ✅ CORRECT ORDER — specific first, broad last
print("\n--- Correct order ---")
def good_order(a, b):
    try:
        return a / b
    except ZeroDivisionError:    # Specific — checked first
        print(f"  ZeroDivisionError: cannot divide {a} by zero")
        return None
    except TypeError:            # Specific — checked second
        print(f"  TypeError: both arguments must be numbers")
        return None
    except Exception as e:       # Broad — always LAST
        print(f"  Unexpected [{type(e).__name__}]: {e}")
        return None

good_order(10, 0)       # ZeroDivisionError
good_order("10", 2)     # TypeError
result = good_order(10, 4)
print(f"  10 / 4 = {result}")  # No exception

print("\nRULE: Specific exceptions BEFORE broad exceptions. Always.")
````

## Part 4 — Writing Custom Exceptions

### Why Create Custom Exceptions?

| Generic (Built-in) | Custom (Your Application) |
|---|---|
| `ValueError: invalid value` | `InsufficientFundsError: Balance too low` |
| `Exception: error occurred` | `StudentNotFoundError: No student ID 999` |
| `KeyError: key not found` | `BookNotAvailableError: 'Clean Code' on loan` |

**Custom exceptions ARE Unit 2 OOP applied to error handling:**

```python
class MyError(Exception):     # Inherit from Exception
    """Raised when something specific goes wrong."""
    pass                       # 'pass' — inheriting is enough
```

That's it. `MyError` now works exactly like any built-in exception.
You can `raise` it, `except` it, and add custom attributes.

````python
# ── 4.1 BASIC CUSTOM EXCEPTION ───────────────────────────────────────────────
# Three steps: Define → Raise → Catch

# ── Step 1: Define ────────────────────────────────────────────────────────────
class InsufficientFundsError(Exception):
    """Raised when a withdrawal exceeds the available balance."""
    pass   # Inheriting from Exception gives us everything we need

class NegativeAmountError(Exception):
    """Raised when a deposit or withdrawal amount is zero or negative."""
    pass

# ── Step 2: Raise ─────────────────────────────────────────────────────────────
def withdraw(balance, amount):
    if amount <= 0:
        raise NegativeAmountError(f"Amount must be positive. Got: £{amount:.2f}.")
    if amount > balance:
        raise InsufficientFundsError(
            f"Cannot withdraw £{amount:.2f}. "
            f"Available balance: £{balance:.2f}."
        )
    return balance - amount

def deposit(balance, amount):
    if amount <= 0:
        raise NegativeAmountError(f"Amount must be positive. Got: £{amount:.2f}.")
    return balance + amount

# ── Step 3: Catch ─────────────────────────────────────────────────────────────
print("=== Bank Account — Custom Exceptions ===\n")
balance = 300.00

operations = [
    ("withdraw", 100.00),
    ("withdraw", 500.00),    # InsufficientFundsError
    ("deposit",  200.00),
    ("deposit",  -50.00),    # NegativeAmountError
    ("withdraw", 250.00),
]

for op, amount in operations:
    try:
        if op == "withdraw":
            balance = withdraw(balance, amount)
            print(f"  Withdrew £{amount:.2f}. New balance: £{balance:.2f}")
        else:
            balance = deposit(balance, amount)
            print(f"  Deposited £{amount:.2f}. New balance: £{balance:.2f}")
    except InsufficientFundsError as e:
        print(f"  InsufficientFundsError: {e}")
    except NegativeAmountError as e:
        print(f"  NegativeAmountError: {e}")
````

A custom exception is just a class that inherits from Exception.

Because it's a class, it can have:
- Attributes (amount, balance, shortfall)
- Methods
- Custom logic in __init__()

This lets you provide detailed information about an error instead of only a text message.

That's why custom exceptions are common in larger applications—they carry useful context that the except block can inspect and act upon.

````python
# ── 4.2 CUSTOM EXCEPTION WITH ATTRIBUTES ─────────────────────────────────────
# Add __init__ with custom attributes.
# The except block can then access detailed information.

class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        self.amount = amount
        self.balance = balance
        super().__init__(
            "Insufficient funds...\n"
            f"Attempted: {self.amount},"
            f"Availabe: {self.balance}"
            )


def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(amount, balance)


def atm():
    try:
        withdraw(100, 250)
    except InsufficientFundsError as e:
        print(e)
atm()
````

- withdraw() knows the values:
    - amount = 250
    - balance = 100
- atm() catches the exception.
- atm() does **not** have local variables named amount or balance.
- Yet it can still print them because the exception carried those values as attributes.
- If you didn't store them in the exception, atm() would only know: `Withdrawal failed` and would have no direct access to the attempted amount or available balance.

That's the key purpose of custom exception attributes: **they carry data from where the error happened to where the error is handled.**

````python
# ── 4.3 CUSTOM EXCEPTION HIERARCHY ──────────────────────────────────────────────────
# For professional applications, build a hierarchy (just like Unit 2 classes).
# A base exception lets callers catch ALL related errors with one clause.

class AppError(Exception):
    pass

# Level 2
class AuthError(AppError):
    pass

class PaymentError(AppError):
    pass

# Level 3 (more specific)
class InvalidPasswordError(AuthError):
    pass

class CardDeclinedError(PaymentError):
    pass

def login(user, password):
    if password != "1234":
        raise InvalidPasswordError("Wrong password")

def pay(amount):
    if amount > 100:
        raise CardDeclinedError("Card limit exceeded")

try:
    login(user="alice", password="wrong") # <- change the password to "1234"
    pay(200)

except AuthError as e:
    print("Authentication problem:", e)

except PaymentError as e:
    print("Payment problem:", e)

except AppError as e:
    print("General app problem:", e)
````

A bigger example:

## Part 5 — File Handling Modes

### Why Files?

Variables live only while the program runs — all data is lost on exit.

Files let you 
- **persist data** between runs, 
- share data between programs, and 
- store logs, reports, and records.

_Your program's variables are like a whiteboard — easy to write on, but erased when class ends. Files are like a notebook — you can close it, come back tomorrow, and the notes are still there._


Every file operation starts with:
```python
file_object = open(filename, mode, encoding='utf-8')
```

- `"filename.txt"` — the path to the file
- `"mode"` — what you intend to do with the file

### File Modes Reference

| Mode | Symbol | Creates File? | Overwrites? | Description |
|---|---|---|---|---|
| Read | `'r'` | No (must exist) | No | Read only — default |
| Write | `'w'` | Yes | **YES ⚠️** | Overwrites existing content! |
| Append | `'a'` | Yes | No | Adds to end of file |
| Read+Write | `'r+'` | No (must exist) | Partial | Read and write |
| Binary Read | `'rb'` | No | No | Images, PDFs, audio |
| Binary Write | `'wb'` | Yes | YES ⚠️ | Binary output |

> **⚠️ Critical:** Mode `'w'` **DELETES all existing content** before writing.
> Use `'a'` when you want to **ADD** to an existing file.

### Always specify encoding

```python
open(filename, 'r', encoding='utf-8')   # ← always add this!
```
Without it, Python uses the OS default — which causes `UnicodeDecodeError` on Windows files with accented characters.

`Encoding` = how text is converted into bytes and back.
- Computers store files as bytes, but humans read text, so Python needs a rule to translate between them.

Common encoding
- UTF-8 (most important and widely used)
    - Supports almost all languages
- Others (less common today):
    - ASCII (very limited)
    - Latin-1 (older Western systems)

| Exception | Cause |

````python
# ── 8.1 HANDLING FILE EXCEPTIONS ─────────────────────────────────────────────
print("=== File Exception Handling ===\n")

# ── FileNotFoundError ─────────────────────────────────────────────────────────
print("--- FileNotFoundError ---")
try:
    with open("demo_files/ghost.txt", "r", encoding="utf-8") as f:
        content = f.read()
except FileNotFoundError as e:
    print(f"  Caught : {e}")
    print("  Action : File doesn't exist — show a friendly message\n")
````

````python
# ── 8.1 HANDLING FILE EXCEPTIONS ─────────────────────────────────────────────
# ── Full safe file loader ─────────────────────────────────────────────────────
print("--- Full Pattern: safe_load_file ---")

def safe_load_file(filepath):
    """Load a text file safely, handling all common file exceptions."""
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        print(f"  File not found   : '{filepath}'")
        return None
    except PermissionError:
        print(f"  Permission denied: '{filepath}'")
        return None
    except UnicodeDecodeError:
        print(f"  Encoding error   : '{filepath}'")
        return None
    except OSError as e:
        print(f"  OS error [{type(e).__name__}]: {e}")
        return None

content = safe_load_file("demo_files/students.txt")
if content:
    lines = content.strip().splitlines()
    print(f"  Loaded {len(lines)} records from students.txt")

safe_load_file("demo_files/nonexistent.txt")
````

## Part 9 — The with Statement

### The Problem Without `with`

```python
# FRAGILE — if an exception occurs between open() and close(), the file stays LOCKED
f = open('students.txt', 'r')
content = f.read()
# ← if an exception occurs here, f.close() NEVER runs!
f.close()
```

### The Solution

```python
# ROBUST — file is ALWAYS closed, even if an exception occurs inside the block
with open('students.txt', 'r', encoding='utf-8') as f:
    content = f.read()
# ← file automatically closed here, no matter what happened
```

The `with` statement calls `f.close()` automatically when the block exits —
**even if an exception is raised inside the block.**

### The Gold Standard Pattern

```python
try:
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            process(line.strip())
except FileNotFoundError:
    print(f'{filename} not found.')
except OSError as e:
    print(f'File error: {e}')
```

Use `with` **inside** `try/except` — best of both worlds.

````python
# ── 9.2 PRACTICAL: Student Grade File Manager ────────────────────────────────
# Combining Unit 3 skills: custom exceptions + file I/O + with + try/except

class GradeFileError(Exception):
    """Base exception for grade file operations."""
    pass

class MalformedGradeDataError(GradeFileError):
    """Raised when a line in the grade file has the wrong format."""
    def __init__(self, line_num, line):
        super().__init__(f"Line {line_num}: malformed data: {line!r}")

def load_grades(filepath):
    """Load student grades from a CSV file."""
    students = []
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            for num, line in enumerate(f, 1):
                line = line.strip()
                if not line:
                    continue
                parts = line.split(",")
                if len(parts) != 3:
                    raise MalformedGradeDataError(num, line)
                name, score_str, result = parts
                students.append({"name": name, "score": int(score_str.strip()),
                                  "result": result.strip()})
    except FileNotFoundError:
        print(f"Grade file '{filepath}' not found.")
    except MalformedGradeDataError as e:
        print(f"Data error: {e}")
    return students

def save_grades(filepath, students):
    """Save student grades to a CSV file."""
    try:
        with open(filepath, "w", encoding="utf-8") as f:
            for s in students:
                f.write(f"{s['name']},{s['score']},{s['result']}\n")
        print(f"Saved {len(students)} records → '{filepath}'")
    except PermissionError as e:
        print(f"Cannot save: {e}")

print("=== Student Grade File Manager ===\n")
grades = load_grades("demo_files/students.txt")
print(f"Loaded {len(grades)} students.")

grades.append({"name": "Isla", "score": 78, "result": "Pass"})
save_grades("demo_files/students_updated.txt", grades)

scores = [s["score"] for s in grades]
passing = sum(1 for s in grades if s["result"] == "Pass")
print(f"\nClass size : {len(grades)}")
print(f"Average    : {sum(scores) / len(scores):.1f}")
print(f"Pass rate  : {passing}/{len(grades)} ({passing / len(grades) * 100:.0f}%)")
````

## Part 10 — Summary and Key Takeaways

### The 7 Golden Rules

1. **Always NAME the exception** — never use bare `except:`
2. **Keep `try` blocks SHORT** — only wrap lines that might fail
3. **Always use `with` for file operations** — never call `.close()` manually
4. **Always specify `encoding='utf-8'`** when opening text files
5. **Use `'a'` not `'w'` when ADDING** to existing files
6. **Create custom exceptions** for domain-specific errors in professional code
7. **Test FAILURE paths** as much as success paths
