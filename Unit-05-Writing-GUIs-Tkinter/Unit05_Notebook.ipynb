## 1. Introduction to Tkinter

### What Is a GUI?

A **Graphical User Interface (GUI)** lets users interact with a program by clicking buttons, typing in text boxes, and selecting options — no terminal required.

### What Is Tkinter?

**Tkinter** is Python's standard, built-in GUI library. It comes with every Python installation — no `pip install` needed.

```python
import tkinter as tk   # conventional alias
```

### Event-Driven Programming

GUI programs are **event-driven**: they start, open a window, and then *wait*. When the user does something (click, type, close), an **event** fires and your **callback function** runs.

```
Window opens
     ↓
Event loop starts (mainloop())
     ↓
User does something   →   Your callback function runs
     ↓
Event loop continues
     ↓
User closes window    →   Program ends
```

> **Analogy:** An event loop is a shop assistant standing at the counter — waiting patiently, then serving when a customer arrives.

````python
# ── Demo: Hello World ────────────────────────────────────────────────────
# This window auto-closes after 2 seconds.
# Copy to a .py file to keep it open for interaction.

import tkinter as tk

# 1. Create the main window (root)
root = tk.Tk()
root.title('Hello World')
root.geometry('400x180') # width x height in pixels

# 2. Add a label widget
label = tk.Label(
    root,
    text='Hello, World!',
    font=('Segoe UI', 24, 'bold'),
    fg='#1b5e20'
)
label.pack(expand=True) # place the label in the window

# 3. Start the event loop — the window stays open until closed
root.mainloop()
# print('Window closed.')
````

## 2. Component and Events

In Tkinter, 
- A `component` (more commonly called a widget) is a graphical user interface (GUI) element, while 
- An `event` is an action that occurs during the program's execution, usually triggered by the user or the system.

**Component**

Common Tkinter widgets (component) include:

| Widget | Class | Purpose |
|---|---|---|
| Label | `tk.Label` | Display static text or an image |
| Button | `tk.Button` | Clickable action trigger |
| Entry | `tk.Entry` | Single-line text input |
| Text | `tk.Text` | Multi-line text area |
| Checkbutton | `tk.Checkbutton` | On/off checkbox |
| Radiobutton | `tk.Radiobutton` | Select one from several options |
| Listbox | `tk.Listbox` | Scrollable list of items |
| Frame | `tk.Frame` | Container for grouping widgets |
| Canvas | `tk.Canvas` | Drawing area |

Every widget follows the same three-step pattern:

```
# Step 1: Create the widget — attach it to a parent window
widget = tk.WidgetClass(parent, option=value, ...)

# Step 2: Configure its appearance and behaviour
widget.config(bg="white", font=("Segoe UI", 12))

# Step 3: Place it in the window using a geometry manager
widget.pack()   # or .grid() or .place()

```

### The Root Window

The **root window** (`tk.Tk()`) is the top-level container for your entire application. Every other widget goes inside it.

### Key configuration methods

| Method | Purpose |
|---|---|
| `root.title('text')` | Title bar text |
| `root.geometry('WxH')` | Initial size in pixels |
| `root.minsize(w, h)` | Minimum resize limit |
| `root.configure(bg='colour')` | Background colour |
| `root.resizable(False, False)` | Prevent resizing |
| `root.protocol('WM_DELETE_WINDOW', fn)` | Intercept close button |

### `root.mainloop()`

- Makes the window **visible**
- Listens continuously for events
- Blocks until the window is closed
- **Must always be the last call in your script**

````python
# ── Demo: Configured root window ────────────────────────────────────────
import tkinter as tk
import tkinter.messagebox as mb

root = tk.Tk()
root.title('Student Portal — BT151CO')
root.geometry('500x300')
root.minsize(250,150)
root.configure(bg='#f5f5f5')
root.resizable(False, True)

tk.Label(
    root,
    text='Welcome to the Student Portal',
    font=('Segoe UI', 18, 'bold'),
    bg='#f5f5f5',
    fg='#1b5e20'
).pack(expand=True)

# root.after(2000, root.destroy)
root.mainloop()
# print('Window closed.')
````

**Full Root Example**

````python
import tkinter as tk


def on_close():
    """Ask for confirmation before closing."""
    import tkinter.messagebox as mb
    if mb.askokcancel("Quit", "Do you really want to quit?"):
        root.destroy()


root = tk.Tk()
root.title("Student Portal — BT151CO")
root.geometry("500x350")
root.configure(bg="#f5f5f5")
root.protocol(name="WM_DELETE_WINDOW", func=on_close)   # intercept the close button

label = tk.Label(
    root,
    text="Welcome to the Student Portal",
    font=("Segoe UI", 16, "bold"),
    bg="#f5f5f5",
    fg="#333333",
)
label.pack(pady=30)

root.mainloop()
````

- `askokcancel`: Ask if operation should proceed; return true if the answer is ok
- `root.protocol(protocol_name, function)` is a Tkinter method used to handle window manager events, such as what happens when the user tries to close the application window.
    - protocol_name: The name of the window manager protocol (usually "WM_DELETE_WINDOW").
    - function: The function to execute when that event occurs.

## 4. Buttons

A `tk.Button` is the primary **action trigger** in a GUI.

### The Golden Rule

> **`command=my_function`** — pass a **reference**, never a call.

```python
# ✅ Correct — function called when clicked
btn = tk.Button(
                root, 
                text='Click', 
                command=my_function
                )

# ❌ Wrong — function called immediately at startup
btn = tk.Button(
                root, 
                text='Click', 
                command=my_function()
                )
```

````python
import tkinter as tk

def button_clicked():
    print("Button clicked!")

root = tk.Tk()
root.title("Button Example")

button = tk.Button(
    root,
    text="Click Me",
    command=button_clicked
)

button.pack(pady=20)

root.mainloop()
````

### Key options

| Option | Description |
|---|---|
| `command` | Callback reference (no parentheses) |
| `state` | `'normal'` or `'disabled'` |
| `cursor` | `'hand2'` shows pointer on hover |
| `relief` | `'raised'`, `'flat'`, `'sunken'`, `'groove'` |


````python
import tkinter as tk

def button_clicked():
    print("Button clicked!")

root = tk.Tk()
root.title("Button Example")

button = tk.Button(
    root,
    text="Click Me",
    command=button_clicked, 
    state= "disabled" # comment this to make it normal, as normal is the default
)

button.pack(pady=20)

root.mainloop()
````

````python
import tkinter as tk

def toggle_button():
    if agree.get() == 1:
        submit_button.config(state="normal")
    else:
        submit_button.config(state="disabled")

def submit_button_callback():
    print(f'Submitted Sucessfully')
    root.destroy()

root = tk.Tk()
root.geometry("500x300")

tos_label = tk.Label(root, text="Terms of Services ....",font=15)
tos_label.pack()
label = tk.Label(root, text="\n Rule 1: Do this, \n Rule 2: Do this. ")
label.pack()

agree = tk.IntVar()

tk.Checkbutton(
    root,
    text="I agree",
    variable=agree,
    command=toggle_button
).pack()

submit_button = tk.Button(
    root,
    text="Submit",
    state="disabled", 
    command=submit_button_callback
)
submit_button.pack()

root.mainloop()
````

### Lambda in a Loop — The Late-Binding Trap

```python
# ❌ All buttons print the LAST value of i (which is 2)
for i in range(3):
    tk.Button(root, text=f'Button {i}',
              command=lambda: print(i)).pack()

# ✅ Capture the current value with a default argument
for i in range(3):
    tk.Button(root, text=f'Button {i}',
              command=lambda val=i: print(val)).pack()
```

**Why?** `lambda: print(i)` captures the *variable* `i`, not its value at creation time. By the time any button is clicked, the loop has finished and `i == 2`.

````python
# ── Demo: Lambda in a loop (WRONG vs CORRECT) ─────────────────────────
import tkinter as tk

root = tk.Tk()
root.title("Lambda Loop Demo")
root.geometry("260x180")
root.configure(bg="white")

for i in range(3):

    # ✅ CORRECT VERSION (captures current value of i)
    tk.Button(
        root,
        text=f"Button {i}",
        command=lambda val=i: print(f"Correct button {val} clicked"),
        bg="#2e7d32",
        fg="white",
        width=16
    ).pack(pady=3)

    ## ❌ WRONG VERSION (all buttons will print 2 when clicked)
    # tk.Button(
    #     root,
    #     text=f"Button {i}",
    #     command=lambda i : print(f"Wrong button {i} clicked"),
    #     bg="#c62828",
    #     fg="white",
    #     width=16
    # ).pack(pady=3)

root.mainloop()
````

### Reading and Writing

```python
# Read the current text
text = entry.get()

# Clear then write
entry.delete(0, tk.END)
entry.insert(0, 'Default text')

# Set focus (move cursor here)
entry.focus()
```

````python
import tkinter as tk

def read_text():
    print(entry.get())  # read current text

def reset_text():
    entry.delete(0, tk.END)         # clear text
    entry.insert(0, "Default text") # write text
    entry.focus()                   # move cursor here

root = tk.Tk()
root.geometry("500x150")
root.title("Entry Read & Write Demo")

entry = tk.Entry(root, width=50)
entry.pack(pady=10)

tk.Button(root, text="Read", command=read_text).pack(pady=10)
tk.Button(root, text="Reset", command=reset_text).pack()

root.mainloop()
````

- `Read button` → prints what you typed using entry.get()
- `Reset button` → clears and writes "Default text"
- `entry.focus()` → puts cursor back in the entry box


Even without entry.focus(), your Entry widget can still work normally because Tkinter automatically decides where the cursor should go when the window opens. For example, if your program has only one Entry box, it will often get focus by default, so you can start typing immediately without calling focus(). However, if there are multiple widgets like several Entry fields and buttons, Tkinter may not always select the one you want, so using entry.focus() ensures the cursor is placed exactly where you want, such as the username field in a login form.

### Validation rule of thumb

> Always call `.strip()` before any other check to remove accidental whitespace.

````python
# ── Demo: Entry with validation ─────────────────────────────────────────
# Enter a name and press 'Submit' or hit Enter.
# Try: empty, spaces only, numbers, valid name.

import tkinter as tk

def submit():
    name = name_entry.get().strip()
    if not name:
        result_label.config(text='⚠️  Name cannot be empty.', fg='red')
        return
    if not name.replace(' ', '').isalpha():
        result_label.config(text='⚠️  Name must contain letters only.', fg='red')
        return
    result_label.config(text=f'✅  Welcome, {name}!', fg='#1b5e20')

root = tk.Tk()
root.title('Entry Validation')
root.geometry('420x220')
root.configure(bg='#f5f5f5')

tk.Label(
        root, 
        text='Enter your name:',
        font=('Segoe UI', 12), 
        bg='#f5f5f5'
        ).pack(pady=(20, 4))

name_entry = tk.Entry(root, font=('Segoe UI', 13), width=28,
                      relief='solid', borderwidth=1)
name_entry.pack(ipady=4)
name_entry.bind('<Return>', lambda e : submit())
name_entry.focus()

tk.Button(
        root, 
        text='Submit', 
        command=submit,
        bg='#1b5e20', fg='white',
        font=('Segoe UI', 12), 
        width=12
        ).pack(pady=10)

# always active but is initially defined as empty string
result_label = tk.Label(root, text='', font=('Segoe UI', 12), bg='#f5f5f5') 
result_label.pack()

root.mainloop()
````

bind() in Tkinter is used to connect an event (like a key press or mouse action) to a function.

- `name_entry` → the Entry widget
- `<Return>` → the Enter key
- `submit()` → function to run when Enter is pressed
- `e` → event object (Tkinter automatically sends it)

### Student Lookup Example

A common pattern: look up a record from a dictionary using an ID typed by the user.

````python
# ── Demo: Student lookup ─────────────────────────────────────────────────
# Type S001, S002, or S003 and press Enter or 'Look Up'.
# Expected: student name and GPA appear below.

import tkinter as tk

STUDENTS = {
    'S001': {'name': 'Alice',   'gpa': 3.8},
    'S002': {'name': 'Bob',     'gpa': 2.9},
    'S003': {'name': 'Charlie', 'gpa': 3.5},
}

def lookup():
    sid = id_entry.get().strip().upper()
    if not sid:
        result.config(text='Please enter a student ID.', fg='red')
        return
    if sid not in STUDENTS:
        result.config(text=f'No record found for: {sid}', fg='red')
        return
    s = STUDENTS[sid]
    result.config(
        text=f"Name: {s['name']}   GPA: {s['gpa']}",
        fg='#1a237e'
    )

root = tk.Tk()
root.title('Student Lookup')
root.geometry('420x220')
root.configure(bg='#f5f5f5')

tk.Label(root, text='Enter Student ID:',
         font=('Segoe UI', 12), bg='#f5f5f5').pack(pady=(20, 4))

id_entry = tk.Entry(root, font=('Segoe UI', 13), width=20,
                    relief='solid', borderwidth=1)
id_entry.pack(ipady=4)
id_entry.bind('<Return>', lambda e: lookup())
id_entry.focus()

tk.Button(root, text='Look Up', command=lookup,
          bg='#1a237e', fg='white',
          font=('Segoe UI', 12), width=12).pack(pady=8)

result = tk.Label(root, text='', font=('Segoe UI', 12), bg='#f5f5f5')
result.pack()

root.mainloop()
````

## 6. Tkinter Variables

Tkinter provides special **variable classes** that link widget values to Python variables.

| Class | Holds | Used with |
|---|---|---|
| `tk.StringVar` | String | Entry, Label |
| `tk.IntVar` | Integer | Spinbox, Radiobutton |
| `tk.BooleanVar` | True / False | Checkbutton |
| `tk.DoubleVar` | Float | Scale |

```python
name_var = tk.StringVar()
name_var.set('Alice')       # write a value
print(name_var.get())       # read the value → 'Alice'

# Link to an Entry — they stay in sync automatically
entry = tk.Entry(root, textvariable=name_var)
```

> When the user types in the Entry, `name_var` updates automatically — and when you call `name_var.set(...)`, the Entry display updates too.

**Problem: Managing an Entry Without StringVar**


````python
import tkinter as tk

root = tk.Tk()
root.title("Without StringVar")
root.geometry("200x200")

# Entry widget
entry = tk.Entry(root)
entry.insert(0, "Initial Value")
entry.pack(pady=5)

# Label widget
label = tk.Label(root, text="Initial Value")
label.pack(pady=5)

label2 = tk.Label(root, text="Initial Value")
label2.pack(pady=5)

def change_name():
    text = entry.get()      # Save the text first
    entry.delete(0, tk.END)
    entry.insert(0, text) # update this 
    label.config(text=text) # but also remember to update this. 
    label2.config(text=text)

button = tk.Button(root, text="Change Name", command=change_name)
button.pack(pady=10)

root.mainloop()
````

- The Entry widget stores the text itself.
- Every time you want to read the text, you must call: `entry.get()`
- Every time you want to change the text, you must first remove the old text and then insert the new one:
    - `entry.delete(0, tk.END)`
    - `entry.insert(0, "Bob")`
- Your program is directly dependent on the widget.
- If multiple widgets need the same value, you have to update each widget manually.

**Solution: Using StringVar**

````python
import tkinter as tk

root = tk.Tk()
root.title("With StringVar")
root.geometry("200x200")

# Shared StringVar
text_var = tk.StringVar(value="Initial Value")

# Entry widget
entry = tk.Entry(root, textvariable=text_var)
entry.pack(pady=5)

# Label widgets
label = tk.Label(
            root, 
            textvariable=text_var, 
            font=("Arial", 11),
            bg="#f4f4f4",
            fg="#1f4e79"
            )
label.pack(pady=5)

label2 = tk.Label(
            root, 
            textvariable=text_var, 
            font=("Arial", 11),
            bg="#f4f4f4",
            fg="#1f4e79")
label2.pack(pady=5)

def change_name():
    # Get the current value
    text = text_var.get()

    # Update the StringVar
    text_var.set(text)

# Button
tk.Button(
    root,
    text="Update Labels",
    command=change_name,
    font=("Arial", 10, "bold"),
    bg="#4CAF50",
    fg="white",
    activebackground="#45a049",
    padx=10,
    pady=5
).pack(pady=20)

root.mainloop()
````

- StringVar acts as a shared variable between your Python code and the Entry widget.
- The Entry no longer stores text independently—it displays whatever is stored in name_var.
- To read the text, simply use: `name_var.get()`
- To change the text, simply use: `name_var.set("Bob")`
- When the user types in the `Entry`, `name_var` updates automatically.
- When your Python code calls `name_var.set(...)`, the Entry updates automatically.
- This automatic two-way synchronization makes the code cleaner, easier to maintain, and especially useful when multiple widgets need to share the same value.

````python
# ── Demo: StringVar two-way binding ────────────────────────────────────
# Typing in the Entry updates the Label below in real time.

import tkinter as tk

root = tk.Tk()
root.title('StringVar Demo')
root.geometry('380x180')
root.configure(bg='#ffffff')

name_var = tk.StringVar()

tk.Label(root, text='Type your name:',
         font=('Segoe UI', 12), bg='#ffffff').pack(pady=(20, 4))

entry = tk.Entry(root, textvariable=name_var,
                 font=('Segoe UI', 13), width=26,
                 relief='solid', borderwidth=1)
entry.pack(ipady=4)
entry.focus()

# The label text updates automatically as you type
preview = tk.Label(root, textvariable=name_var,
                   font=('Segoe UI', 14, 'italic'),
                   fg='#1b5e20', bg='#ffffff')
preview.pack(pady=10)


root.mainloop()
print('Done.')
````

## 7. Text Widgets

A `tk.Text` widget provides a **multi-line, scrollable text area** — like a mini word processor inside your window.

- This widget can be used for a variety of applications where the multiline text is required such as: 
    - messaging, 
    - sending information or 
    - displaying information and many other tasks. 
- We can insert media files such as images and links also in the Textwidget.

**Syntax:**
```python
T = Text(root, bg, fg, bd, height, width, font, ..)
```
Optional parameters: 
- `root` → root window
- `bg` → background colour
- `fg` → foreground colour
- `bd` → border of widget
- `height` → height of the widget
- `width` → width of the widget
- `font` → font type of the text
- `cursor` → type of cursor to be used
- `insertontime` → time in milliseconds for which cursor blinking is on
- `insetofftime` → time in milliseconds for which cursor blinking is off
- `padx` → horizontal padding
- `pady` → vertical padding
- `state` → defines if widget is responsive to mouse or keyboard input
- `highlightthickness` → thickness of focus highlight
- `insertwidth` → width of insertion cursor
- `relief` → border style (SUNKEN, RAISED, GROOVE, RIDGE)
- `yscrollcommand` → makes widget vertically scrollable
- `xscrollcommand` → makes widget horizontally scrollable


Some Common Methods
- `index(index)` - To get the specified index.  
- `insert(index)` - To insert a string at a specified index.  
- `see(index)` - Checks if a string is visible or not at a given index.  
- `get(startindex, endindex)` - To get characters within a given range.  
- `delete(startindex, endindex)` - Deletes characters within a specified range.  

Tag Handling Methods
- `tag_delete(tagname)` - To delete a given tag.  
- `tag_add(tagname, startindex, endindex)` - To tag the string in the specified range.  
- `tag_remove(tagname, startindex, endindex)` - To remove a tag from a specified range.  

Mark Handling Methods
- `mark_names()` - To get all the marks in the given range.  
- `index(mark)` - To get index of a mark.  
- `mark_gravity()` - To get the gravity of a given mark.



**Common Use Cases**

**2. Chat Applications**

**3. Forms and Input Fields**

**4. Log or Console Viewers**
Used in GUI-based log monitors or terminals.
- Display logs with `insert()`

**5. Educational Tools**
Used in:
- coding practice apps  
- interactive tutorials  
- syntax learning environments  

### Position format: `'line.column'`

- `'1.0'` = line 1, column 0 (the very start)
- `tk.END` = the position after the last character

```python
text.insert('1.0', 'Hello!')      # insert at start
text.insert(tk.END, '\nNew line') # append at end
content = text.get('1.0', tk.END) # read everything
text.delete('1.0', tk.END)        # clear everything
text.see(tk.END)                  # scroll to the bottom
```

- `"1.0"` → start of text (line 1, column 0)
- `"end"` or `tk.END` → end of text (includes trailing newline)
- `"end-1c"` → end minus 1 character (removes extra newline)
- `"insert"` → current cursor position
- `"insert linestart"` → start of current line
- `"insert lineend"` → end of current line
- `"sel.first"` → start of selected text
- `"sel.last"` → end of selected text
- `"line.column"` → format for direct indexing

Examples:
- `"2.0"` → line 2, column 0
- `"5.10"` → line 5, column 10

Relative movement:
- `"insert+1c"` → move forward 1 character
- `"insert-1c"` → move backward 1 character
- `"insert+1l"` → move down 1 line
- `"insert-1l"` → move up 1 line
- `"insert wordstart"` → start of current word
- `"insert wordend"` → end of current word

Line shortcuts:
- `"insert linestart"` → start of current line
- `"insert lineend"` → end of current line

Special:
- `"1.0"` → beginning of text
- `"end"` → end of text
- `"end-1c"` → last real character
- `"insert"` → cursor position

Most important:
- `"1.0"` → start
- `"end"` → end
- `"end-1c"` → real end (no extra newline)
- `"insert"` → cursor
- `"sel.first"` → selection start
- `"sel.last"` → selection end

````python
import tkinter as tk

root = tk.Tk()

text = tk.Text(root, height=8, width=40)
text.pack()

text.insert("1.0", "Hello World\nThis is Tkinter\nEnjoy coding!")

# 1. "1.0" → start of text
print(text.get("1.0", "1.5"))
# Output: "Hello"

# 2. "end" → end of text (includes extra newline)
print(text.get("1.0", "end"))

# 3. "end-1c" → real end (no extra newline)
print(text.get("1.0", "end-1c"))

# 4. "insert" → current cursor position
text.mark_set("insert", "1.6")  # move cursor after "Hello"
text.insert("insert", "!!!")    # inserts at cursor

# 5. "sel.first" and "sel.last" → selection
text.tag_add("sel", "1.0", "1.5")  # select "Hello"
print(text.get("sel.first", "sel.last"))
# Output: "Hello"

# 6. "insert linestart" → start of current line
print(text.get("insert linestart", "insert"))

# 7. "insert lineend" → end of current line
print(text.get("insert", "insert lineend"))

root.mainloop()
````

## 9. Layout Management

Tkinter has three **geometry managers** for placing widgets:

| Manager | Best for | Key options |
|---|---|---|
| `pack()` | Linear stacking | `side`, `fill`, `expand`, `padx/pady` |
| `grid()` | Forms and tables | `row`, `column`, `sticky`, `columnspan` |
| `place()` | Exact pixel control | `x`, `y`, `width`, `height` |

> **Critical rule:** Never mix `pack()` and `grid()` **in the same container** — it causes a `TclError`. Use `Frame`s to separate regions.

### `pack()` key options

```python
widget.pack(side='left', fill='x', expand=True, padx=5, pady=5)
```

### `grid()` key options

```python
widget.grid(row=0, column=1, sticky='ew', padx=8, columnspan=2)
```

### Using `Frame` for structure

```python
header  = tk.Frame(root, bg='#1b5e20', height=60)
content = tk.Frame(root, bg='#ffffff')
footer  = tk.Frame(root, bg='#f5f5f5', height=30)

header.pack(fill='x')
content.pack(fill='both', expand=True)
footer.pack(fill='x')
# Each frame can use pack() or grid() independently
```

````python
# ── Demo: pack() vs grid() comparison ──────────────────────────────────
# Two separate windows demonstrating each manager.

import tkinter as tk

# ── Window 1: pack() ─────────────────────────────────────────────────────
root1 = tk.Tk()
root1.title('pack() demo')
root1.geometry('300x240')
root1.configure(bg='#ffffff')

header = tk.Frame(root1, bg='#1b5e20', height=40)
header.pack(fill='x')
header.pack_propagate(False)
tk.Label(header, text='Header (pack)', bg='#1b5e20', fg='white',
         font=('Segoe UI', 12, 'bold')).pack(expand=True)

body = tk.Frame(root1, bg='#ffffff')
body.pack(fill='both', expand=True, padx=10, pady=10)
for text in ['Item A', 'Item B', 'Item C']:
    tk.Label(body, text=text, font=('Segoe UI', 12),
             bg='#ffffff').pack(anchor='w', pady=4)

btn_frame = tk.Frame(root1, bg='#ffffff')
btn_frame.pack(pady=6)
tk.Button(btn_frame, text='OK', bg='#1b5e20', fg='white', width=8).pack(
    side='left', padx=4)
tk.Button(btn_frame, text='Cancel', bg='#c62828', fg='white', width=8).pack(
    side='left', padx=4)

root1.after(3000, root1.destroy)

# ── Window 2: grid() ─────────────────────────────────────────────────────
root2 = tk.Tk()
root2.title('grid() demo — form layout')
root2.geometry('360x200')
root2.configure(bg='#f5f5f5')

fields = [('Name:', 0), ('Student ID:', 1), ('Score:', 2)]
for label_text, row in fields:
    tk.Label(root2, text=label_text, font=('Segoe UI', 12),
             bg='#f5f5f5').grid(row=row, column=0, sticky='e',
                                padx=10, pady=6)
    tk.Entry(root2, font=('Segoe UI', 12), width=20,
             relief='solid', borderwidth=1).grid(row=row, column=1,
                                                  padx=10, pady=6, ipady=3)

tk.Button(root2, text='Submit', bg='#1a237e', fg='white',
          font=('Segoe UI', 12)).grid(row=3, column=0, columnspan=2, pady=10)

root2.after(3000, root2.destroy)

root1.mainloop()
print('Done.')
````

### Practice Task 1 — Login Form

Build a login form with:
- A **Username** Entry field
- A **Password** Entry field (masked with `show='*'`)

**Bonus:** Disable the Login button after 3 failed attempts and show a lockout message.

````python
# ── Practice Task 1 — Solution ──────────────────────────────────────────
import tkinter as tk

VALID_USER = 'admin'
VALID_PASS = 'secret123'
MAX_ATTEMPTS = 3

attempt_count = 0

def login():
    global attempt_count
    username = user_entry.get().strip()
    password = pass_entry.get()

    if username == VALID_USER and password == VALID_PASS:
        status_label.config(text='✅ Login successful!', fg='#1b5e20')
        login_btn.config(state='disabled')
    else:
        attempt_count += 1
        remaining = MAX_ATTEMPTS - attempt_count
        pass_entry.delete(0, tk.END)
        if remaining > 0:
            status_label.config(
                text=f'❌ Invalid credentials. {remaining} attempt(s) left.',
                fg='red'
            )
        else:
            status_label.config(
                text='🔒 Account locked. Too many failed attempts.',
                fg='red'
            )
            login_btn.config(state='disabled')

root = tk.Tk()
root.title('Login')
root.geometry('400x260')
root.configure(bg='#f5f5f5')
root.resizable(False, False)

tk.Label(root, text='Login', font=('Segoe UI', 18, 'bold'),
         bg='#f5f5f5', fg='#1b5e20').pack(pady=(20, 14))

form = tk.Frame(root, bg='#f5f5f5')
form.pack()

tk.Label(form, text='Username:', font=('Segoe UI', 12),
         bg='#f5f5f5').grid(row=0, column=0, sticky='e', padx=8, pady=4)
user_entry = tk.Entry(form, font=('Segoe UI', 12), width=22,
                      relief='solid', borderwidth=1)
user_entry.grid(row=0, column=1, ipady=4, pady=4)

tk.Label(form, text='Password:', font=('Segoe UI', 12),
         bg='#f5f5f5').grid(row=1, column=0, sticky='e', padx=8, pady=4)
pass_entry = tk.Entry(form, font=('Segoe UI', 12), width=22,
                      show='*', relief='solid', borderwidth=1)
pass_entry.grid(row=1, column=1, ipady=4, pady=4)
pass_entry.bind('<Return>', lambda e: login())

login_btn = tk.Button(root, text='Login', command=login,
                      bg='#1b5e20', fg='white',
                      font=('Segoe UI', 12), width=14)
login_btn.pack(pady=12)

status_label = tk.Label(root, text='', font=('Segoe UI', 11), bg='#f5f5f5')
status_label.pack()

user_entry.focus()
root.after(6000, root.destroy)
root.mainloop()
print('Done.')
````

- `F = (C × 9/5) + 32`
- `C = (F − 32) × 5/9`

````python
# ── Practice Task 2 — Solution ──────────────────────────────────────────
import tkinter as tk

def convert():
    raw = temp_entry.get().strip()
    if not raw:
        result_label.config(text='⚠️ Enter a temperature.', fg='red'); return
    try:
        value = float(raw)
    except ValueError:
        result_label.config(text='⚠️ Must be a number.', fg='red'); return

    if direction.get() == 'c2f':
        converted = (value * 9 / 5) + 32
        result_label.config(
            text=f'{value}°C  =  {converted:.1f}°F', fg='#1b5e20')
    else:
        converted = (value - 32) * 5 / 9
        result_label.config(
            text=f'{value}°F  =  {converted:.1f}°C', fg='#1b5e20')

root = tk.Tk()
root.title('Temperature Converter')
root.geometry('380x250')
root.configure(bg='#ffffff')

tk.Label(root, text='Temperature Converter',
         font=('Segoe UI', 16, 'bold'), bg='#ffffff',
         fg='#1b5e20').pack(pady=(16, 10))

tk.Label(root, text='Enter temperature:',
         font=('Segoe UI', 11), bg='#ffffff').pack()
temp_entry = tk.Entry(root, font=('Segoe UI', 13), width=16,
                      relief='solid', borderwidth=1)
temp_entry.pack(ipady=4, pady=4)
temp_entry.bind('<Return>', lambda e: convert())

direction = tk.StringVar(value='c2f')
radio_frame = tk.Frame(root, bg='#ffffff')
radio_frame.pack(pady=6)
tk.Radiobutton(radio_frame, text='Celsius → Fahrenheit',
               variable=direction, value='c2f',
               font=('Segoe UI', 11), bg='#ffffff').pack(side='left', padx=6)
tk.Radiobutton(radio_frame, text='Fahrenheit → Celsius',
               variable=direction, value='f2c',
               font=('Segoe UI', 11), bg='#ffffff').pack(side='left', padx=6)

tk.Button(root, text='Convert', command=convert,
          bg='#1b5e20', fg='white',
          font=('Segoe UI', 12), width=12).pack(pady=8)

result_label = tk.Label(root, text='', font=('Segoe UI', 13, 'bold'),
                        bg='#ffffff')
result_label.pack()

temp_entry.focus()
root.after(6000, root.destroy)
root.mainloop()
print('Done.')
````

### Practice Task 3 — Notes Editor

Build a simple notes app:
- A `Text` widget with a vertical scrollbar
- A **Word Count** button that shows the current word count in a label
- A **Clear** button that clears all text (with confirmation via `messagebox.askokcancel`)
- The Text widget should wrap at word boundaries

````python
# ── Practice Task 3 — Solution ──────────────────────────────────────────
import tkinter as tk
import tkinter.messagebox as mb

def word_count():
    content = text_area.get('1.0', tk.END).strip()
    words = len(content.split()) if content else 0
    count_label.config(text=f'Words: {words}')

def clear_text():
    if mb.askokcancel('Clear', 'Clear all text?'):
        text_area.delete('1.0', tk.END)
        count_label.config(text='Words: 0')

root = tk.Tk()
root.title('Notes Editor')
root.geometry('540x380')
root.configure(bg='#ffffff')

tk.Label(root, text='Notes', font=('Segoe UI', 15, 'bold'),
         bg='#ffffff', fg='#1b5e20').pack(pady=(12, 4))

# Text area + scrollbar in a Frame
text_frame = tk.Frame(root)
text_frame.pack(padx=14, fill='both', expand=True)

scrollbar = tk.Scrollbar(text_frame)
scrollbar.pack(side='right', fill='y')

text_area = tk.Text(text_frame, font=('Segoe UI', 12), wrap='word',
                    yscrollcommand=scrollbar.set,
                    relief='solid', borderwidth=1, undo=True)
text_area.pack(side='left', fill='both', expand=True)
scrollbar.config(command=text_area.yview)

# Bottom controls
ctrl = tk.Frame(root, bg='#ffffff')
ctrl.pack(pady=8)

tk.Button(ctrl, text='Word Count', command=word_count,
          bg='#1b5e20', fg='white', font=('Segoe UI', 11),
          width=12).pack(side='left', padx=5)
tk.Button(ctrl, text='Clear', command=clear_text,
          bg='#c62828', fg='white', font=('Segoe UI', 11),
          width=8).pack(side='left', padx=5)
count_label = tk.Label(ctrl, text='Words: 0', font=('Segoe UI', 11),
                       bg='#ffffff')
count_label.pack(side='left', padx=10)

text_area.focus()
root.after(5000, root.destroy)
root.mainloop()
print('Done.')
````

### Mini Challenge 1 — Student Grade Calculator

Build a complete GUI application:

**Inputs:**
- Student Name (Entry)
- Student ID (Entry)
- Comma-separated scores (Entry, e.g. `85, 72, 91`)
- Honours Programme checkbox (+5 bonus to average, capped at 100)

**Processing:**
- Validate all fields (not empty, scores are numbers between 0–100)
- Calculate average
- Assign letter grade: A ≥ 90, B ≥ 75, C ≥ 60, D ≥ 50, F < 50
- Show PASS / FAIL based on average ≥ 50

**Output:** A formatted report in a read-only `Text` widget with colour-coded lines

**Extra:** A Clear button that resets all fields and the report

````python
# ── Mini Challenge 1 — Student Grade Calculator ─────────────────────────
# Run this as a .py file for full interaction.
# In Jupyter it auto-closes after 8 seconds.

import tkinter as tk
import tkinter.messagebox as mb

def calculate_grade(average):
    """Return letter grade for a numeric average."""
    if average >= 90: return 'A'
    if average >= 75: return 'B'
    if average >= 60: return 'C'
    if average >= 50: return 'D'
    return 'F'

def calculate():
    name = name_entry.get().strip()
    sid  = id_entry.get().strip().upper()
    raw  = scores_entry.get().strip()

    if not name:
        mb.showwarning('Input Error', 'Name cannot be empty.'); return
    if not sid:
        mb.showwarning('Input Error', 'Student ID cannot be empty.'); return
    if not raw:
        mb.showwarning('Input Error', 'Enter at least one score.'); return

    try:
        scores = [float(s.strip()) for s in raw.split(',')]
    except ValueError:
        mb.showerror('Input Error', 'Scores must be numbers, e.g. 85, 72, 91')
        return

    for s in scores:
        if not (0 <= s <= 100):
            mb.showerror('Input Error', f'Score {s} is out of range (0-100).')
            return

    average = sum(scores) / len(scores)
    if honours_var.get():
        average = min(average + 5, 100)

    grade  = calculate_grade(average)
    passed = average >= 50

    report.config(state='normal')
    report.delete('1.0', tk.END)
    report.insert(tk.END, '=' * 42 + '\n', 'border')
    report.insert(tk.END, '  GRADE REPORT\n', 'header')
    report.insert(tk.END, '=' * 42 + '\n', 'border')
    report.insert(tk.END, f'  Name:    {name}\n')
    report.insert(tk.END, f'  ID:      {sid}\n')
    scores_str = ', '.join(
        str(int(s)) if s == int(s) else str(s) for s in scores
    )
    report.insert(tk.END, f'  Scores:  {scores_str}\n')
    report.insert(tk.END, f'  Average: {average:.1f}\n')
    report.insert(tk.END, f'  Grade:   {grade}\n')
    report.insert(tk.END, f'  Honours: {"Yes" if honours_var.get() else "No"}\n')
    report.insert(tk.END, '-' * 42 + '\n', 'border')
    report.insert(tk.END,
        '  RESULT: PASS ✅\n' if passed else '  RESULT: FAIL ❌\n',
        'pass' if passed else 'fail'
    )
    report.insert(tk.END, '=' * 42 + '\n', 'border')
    report.config(state='disabled')

def clear_all():
    for e in (name_entry, id_entry, scores_entry):
        e.delete(0, tk.END)
    honours_var.set(False)
    report.config(state='normal')
    report.delete('1.0', tk.END)
    report.config(state='disabled')
    name_entry.focus()

# ── Window ────────────────────────────────────────────────────────────────
root = tk.Tk()
root.title('Student Grade Calculator')
root.geometry('500x570')
root.configure(bg='#ffffff')
root.resizable(False, False)

# Header
header = tk.Frame(root, bg='#1b5e20', height=55)
header.pack(fill='x')
header.pack_propagate(False)
tk.Label(header, text='Student Grade Calculator',
         font=('Segoe UI', 16, 'bold'),
         bg='#1b5e20', fg='white').pack(expand=True)

# Form
form = tk.Frame(root, bg='#ffffff', padx=28, pady=16)
form.pack(fill='x')

def _lbl(text):
    return tk.Label(form, text=text, font=('Segoe UI', 11),
                    bg='#ffffff', anchor='w')
def _ent():
    return tk.Entry(form, font=('Segoe UI', 12),
                    relief='solid', borderwidth=1, width=30)

_lbl('Student Name:').grid(row=0, column=0, sticky='w', pady=(0, 2))
name_entry = _ent(); name_entry.grid(row=1, column=0, ipady=4, pady=(0, 10))

_lbl('Student ID:').grid(row=2, column=0, sticky='w', pady=(0, 2))
id_entry = _ent(); id_entry.grid(row=3, column=0, ipady=4, pady=(0, 10))

_lbl('Scores (comma-separated):').grid(row=4, column=0, sticky='w', pady=(0, 2))
scores_entry = _ent(); scores_entry.grid(row=5, column=0, ipady=4, pady=(0, 10))

honours_var = tk.BooleanVar()
tk.Checkbutton(form, text='Honours Programme  (+5 bonus)',
               variable=honours_var,
               font=('Segoe UI', 11), bg='#ffffff',
               activebackground='#ffffff',
               cursor='hand2').grid(row=6, column=0, sticky='w', pady=(0, 14))

# Buttons
btn_frame = tk.Frame(root, bg='#ffffff')
btn_frame.pack()
tk.Button(btn_frame, text='Calculate',
          command=calculate, bg='#1b5e20', fg='white',
          font=('Segoe UI', 12), width=13, cursor='hand2'
          ).pack(side='left', padx=5)
tk.Button(btn_frame, text='Clear',
          command=clear_all, bg='#c62828', fg='white',
          font=('Segoe UI', 12), width=9, cursor='hand2'
          ).pack(side='left', padx=5)

# Report
tk.Label(root, text='Report:', font=('Segoe UI', 11),
         bg='#ffffff', anchor='w').pack(padx=28, anchor='w', pady=(14, 2))

report = tk.Text(root, width=50, height=10,
                 font=('Courier New', 11), state='disabled',
                 relief='solid', borderwidth=1, bg='#fafafa')
report.tag_configure('header', foreground='#1b5e20',
                     font=('Courier New', 12, 'bold'))
report.tag_configure('border', foreground='#9e9e9e')
report.tag_configure('pass',   foreground='#1b5e20',
                     font=('Courier New', 12, 'bold'))
report.tag_configure('fail',   foreground='#b71c1c',
                     font=('Courier New', 12, 'bold'))
report.pack(padx=28, pady=(0, 16))

name_entry.focus()
root.after(8000, root.destroy)
root.mainloop()
print('Done.')
````

### Mini Challenge 2 — Library Book Tracker

Build a library book tracking application:

**Features:**
- Entry fields for **Title** and **Author**
- A **Status** `Checkbutton`: ticked = Returned, unticked = On Loan
- An **Add Book** button that appends the book to a `Text` widget log
- Each log entry should be colour-coded: green for Returned, orange for On Loan
- A **Clear Log** button
- A **Summary** button that shows: total books, returned count, on-loan count

**Data:** store books in a list of dicts: `{'title': ..., 'author': ..., 'returned': True/False}`

````python
# ── Mini Challenge 2 — Library Book Tracker ─────────────────────────────
# Run as a .py file for full interaction. Auto-closes in 8 s in Jupyter.

import tkinter as tk
import tkinter.messagebox as mb

books = []   # list of {title, author, returned}

def add_book():
    title  = title_entry.get().strip()
    author = author_entry.get().strip()
    if not title or not author:
        mb.showwarning('Input Error', 'Both Title and Author are required.')
        return
    returned = returned_var.get()
    books.append({'title': title, 'author': author, 'returned': returned})

    status_text = '✅ Returned' if returned else '📤 On Loan'
    tag = 'returned' if returned else 'loan'
    log.config(state='normal')
    log.insert(tk.END, f'  {title} — {author}  [{status_text}]\n', tag)
    log.see(tk.END)
    log.config(state='disabled')

    title_entry.delete(0, tk.END)
    author_entry.delete(0, tk.END)
    returned_var.set(False)
    title_entry.focus()

def clear_log():
    if not mb.askokcancel('Clear', 'Clear all records?'): return
    books.clear()
    log.config(state='normal')
    log.delete('1.0', tk.END)
    log.config(state='disabled')

def show_summary():
    total    = len(books)
    returned = sum(1 for b in books if b['returned'])
    on_loan  = total - returned
    mb.showinfo('Summary',
        f'Total books:  {total}\n'
        f'Returned:     {returned}\n'
        f'On Loan:      {on_loan}')

root = tk.Tk()
root.title('Library Book Tracker')
root.geometry('520x450')
root.configure(bg='#f5f5f5')

# Header
hdr = tk.Frame(root, bg='#1b5e20', height=50)
hdr.pack(fill='x')
hdr.pack_propagate(False)
tk.Label(hdr, text='Library Book Tracker',
         font=('Segoe UI', 15, 'bold'),
         bg='#1b5e20', fg='white').pack(expand=True)

# Form
form = tk.Frame(root, bg='#f5f5f5', padx=20, pady=10)
form.pack(fill='x')

tk.Label(form, text='Title:', font=('Segoe UI', 11),
         bg='#f5f5f5').grid(row=0, column=0, sticky='e', padx=6, pady=4)
title_entry = tk.Entry(form, font=('Segoe UI', 12), width=28,
                       relief='solid', borderwidth=1)
title_entry.grid(row=0, column=1, ipady=4, pady=4)

tk.Label(form, text='Author:', font=('Segoe UI', 11),
         bg='#f5f5f5').grid(row=1, column=0, sticky='e', padx=6, pady=4)
author_entry = tk.Entry(form, font=('Segoe UI', 12), width=28,
                        relief='solid', borderwidth=1)
author_entry.grid(row=1, column=1, ipady=4, pady=4)

returned_var = tk.BooleanVar()
tk.Checkbutton(form, text='Returned (untick = On Loan)',
               variable=returned_var,
               font=('Segoe UI', 11), bg='#f5f5f5',
               activebackground='#f5f5f5'
               ).grid(row=2, column=1, sticky='w', pady=4)

# Buttons
btn_row = tk.Frame(root, bg='#f5f5f5')
btn_row.pack(pady=4)
tk.Button(btn_row, text='Add Book', command=add_book,
          bg='#1b5e20', fg='white', font=('Segoe UI', 11), width=11
          ).pack(side='left', padx=4)
tk.Button(btn_row, text='Summary', command=show_summary,
          bg='#1a237e', fg='white', font=('Segoe UI', 11), width=10
          ).pack(side='left', padx=4)
tk.Button(btn_row, text='Clear Log', command=clear_log,
          bg='#c62828', fg='white', font=('Segoe UI', 11), width=10
          ).pack(side='left', padx=4)

# Log
log = tk.Text(root, state='disabled', font=('Segoe UI', 11),
              height=12, relief='solid', borderwidth=1, bg='#fafafa')
log.tag_configure('returned', foreground='#1b5e20')
log.tag_configure('loan',     foreground='#e65100')
log.pack(padx=16, pady=8, fill='both', expand=True)

title_entry.focus()
root.after(8000, root.destroy)
root.mainloop()
print('Done.')
````

## 13. Summary

### What you learned in Unit 5

| Topic | Key point |
|---|---|
| GUI | A visual interface — no terminal required |
| Tkinter | Python's built-in GUI library — `import tkinter as tk` |
| Event loop | `root.mainloop()` — always the last line |
| Widgets | Every visual element: Label, Button, Entry, Text, Checkbutton |
| Callbacks | `command=my_func` — pass a reference, never a call |

| Layout | `pack()` for linear; `grid()` for forms; never mix in same container |
| Frame | Invisible container — the key to complex layouts |

### Quick Reference — Common Mistakes

| Mistake | Fix |
|---|---|
| `command=my_func()` | `command=my_func` |
| Mix `pack()` and `grid()` | Use `Frame`s to separate |
| Forget `mainloop()` | Add as last line |
| No `.strip()` on Entry | Always strip first |
| Lambda loop bug | `lambda val=i: f(val)` |
| User edits report Text | `state='disabled'` |
