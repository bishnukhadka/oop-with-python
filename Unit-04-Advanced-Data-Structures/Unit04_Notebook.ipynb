## Section 1 — Comprehensions

### 1.1 What Is a Comprehension?

A **comprehension** is a compact, single-line construct for building a new
collection by **transforming** and/or **filtering** an existing iterable.

The idea comes from mathematical set-builder notation:

```
{ x² | x ∈ {1,2,3,4,5} }   ← mathematics
[x**2 for x in range(1, 6)] ← Python list comprehension
```

**Syntax:**
```
[ expression   for   item   in   iterable   if   condition ]
    ↑               ↑           ↑                ↑
  What to put    Variable     Source          Optional filter
  in the list
```

Run the cells below to see comprehensions in action.


````python
# Traditional for-loop approach
scores = [45, 82, 67, 30, 91, 55, 73, 48]

passing_loop = []
for score in scores:
    if score >= 50:
        passing_loop.append(score)

print("For-loop result:", passing_loop)

# List comprehension — same result, one line
passing_comp = [score for score in scores if score >= 50]
print("Comprehension:  ", passing_comp)

# Both produce the same output:
# [82, 67, 91, 55, 73]
````

````python
# More list comprehension examples

# 1. Square every number from 1 to 10
squares = [x**2 for x in range(1, 11)]
print("Squares 1-10:", squares)

# 2. Convert Celsius to Fahrenheit
celsius = [0, 20, 37, 100]
fahrenheit = [(c * 9/5) + 32 for c in celsius]
print("Fahrenheit:  ", fahrenheit)

# 3. Capitalise a list of student names
names = ["alice", "bob", "charlie", "diana"]
proper = [name.capitalize() for name in names]
print("Proper names:", proper)

# 4. Filter — keep only items in stock
inventory = [("Apples", 5), ("Bread", 0), ("Milk", 3), ("Cheese", 0)]
in_stock = [item for item, qty in inventory if qty > 0]
print("In stock:    ", in_stock)
````

### `if`/`else` in Comprehensions

⚠️ There are **two different positions** for `if`. And they have very different meanings:

| Position | Effect | Example |
|---|---|---|
| **End** `[... for x in xs if cond]` | **Filter** — removes non-matching items | `[x for x in xs if x > 0]` |
| **Expression** `[a if cond else b for x in xs]` | **Transform** — chooses a value for every item | `["Y" if x>0 else "N" for x in xs]` |


````python
scores = [45, 82, 67, 30, 91, 55]

# FILTER — only passing scores are included
passing = [s for s in scores if s >= 50]
print("Passing scores (filter):    ", passing)
# [82, 67, 91, 55]

# TRANSFORM — every score gets a label (no items removed)
labels = ["Pass" if s >= 50 else "Fail" for s in scores]
print("All labels (transform):     ", labels)
# ['Fail', 'Pass', 'Pass', 'Fail', 'Pass', 'Pass']
````

````python
scores = [45, 82, 67, 30, 91, 55, 'A', 'A']

# Both together — transform AND filter
boosted = ["Pass" if p >= 50 else "Fail" for p in scores if isinstance(p, (int, float))]
print("New collection:     ", boosted)
````

This is equivalent to doing the following: 

````python
scores = [45, 82, 67, 30, 91, 55, 'A', 'A']
result = [] # container to store the result

for p in scores:
    if isinstance(p, (int, float)):  # filter
        if p >= 50:                  # transform
            result.append("Pass")
        else:
            result.append("Fail")

result
````

### 1.2 Nested List Comprehensions

A **nested comprehension** has two `for` clauses and mirrors the structure
of a nested `for` loop. It is most useful for **flattening** or **building** 2-D data.


````python
# ── Flattening a 2-D list ──────────────────────────────────────────────────
# Exam scores: rows = students, columns = subjects
score_matrix = [
    [85, 72, 91],   # Alice: Maths, English, Science
    [60, 88, 75],   # Bob
    [45, 62, 70],   # Charlie
]

# Flatten: one flat list of ALL scores
all_scores = [score for row in score_matrix for score in row]
print("All scores (flattened):", all_scores)
# [85, 72, 91, 60, 88, 75, 45, 62, 70]

# Equivalent traditional nested loop:
flat_loop = []
for row in score_matrix:
    for score in row:
        flat_loop.append(score)
print("Via nested loop:       ", flat_loop)
````

````python
# ── Building a 2-D structure ───────────────────────────────────────────────
# 5×5 multiplication table
table = [
            [
                row * col 
                for col in range(1, 6)
            ] 
            for row in range(1, 6) 
            # if row > 3 # filter/condition    ### <- Uncomment this
        ]

table
````

````python
table = [
    [
        "Even" if (row * col) % 2 == 0 else "Odd"
        for col in range(1, 6)
        if col > 2
    ]
    for row in range(1, 6)
]
table
````

### 1.3 Dictionary Comprehensions

**Syntax:**
```
{ key_expression : value_expression   for   item   in   iterable   if   condition }
```

Dictionary comprehensions build a new `dict` directly from an iterable.


````python
# ── Building a grade book ──────────────────────────────────────────────────
names  = ["Alice", "Bob", "Charlie", "Diana"]
scores = [85, 42, 91, 67]

grade_book = {name: score for name, score in zip(names, scores)}
print("Grade book:", grade_book)
````

````python
# ── Filtering — passing students only ──────────────────────────────────────
passing_book = {name: score for name, score in zip(names, scores) if score >= 50}
print("Passing:   ", passing_book)
````

````python
# ── Transforming values — convert to letter grades ─────────────────────────
def to_grade(score):
    if score >= 90: return "A"
    if score >= 75: return "B"
    if score >= 60: return "C"
    if score >= 50: return "D"
    return "F"

grade_letters = {name: to_grade(s) for name, s in grade_book.items()}
print("Letters:   ", grade_letters)
````

````python
# ── Swapping keys and values ────────────────────────────────────────────────
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print("Inverted:  ", inverted)
````

````python
# ── Set comprehension (bonus) ──────────────────────────────────────────────
# Set comprehensions use {} without a colon — duplicates are removed automatically
scores_with_dups = [85, 72, 85, 91, 72, 55]
unique_scores = {score for score in scores_with_dups}
print("Unique scores:", unique_scores)   # order may vary

# ── Generator expression (memory-efficient) ────────────────────────────────
# Use () instead of [] for a lazy generator — values computed on demand
gen = (score**2 for score in scores_with_dups)
print("Generator:", list(gen))  # convert to list to see values
````

## Section 2 — Functions: Advanced Features

### 2.1 Default Parameters

A **default parameter** has a pre-assigned value used when the caller
does not supply that argument.

```python
def function_name(param1, param2=default_value):
    ...
```

Parameters with defaults must come **after** all required parameters.


````python
# ── Default parameters in action ───────────────────────────────────────────
def greet_student(name, role="Student", year=1):
    """Greet a student with optional role and year."""
    print(f"Welcome, {name}! Role: {role}. Year {year}.")

# All defaults
greet_student("Alice")

# Override role only
greet_student("Bob", "Tutor")

# Override year using a keyword argument (order-independent)
greet_student("Charlie", year=3)

# Override both
greet_student("Diana", "TA", 2)
````

````python
# ── Shopping total with optional discount ──────────────────────────────────
def calculate_total(items, discount_percent=0, tax_rate=0.20):
    """Calculate shopping total with optional discount and fixed tax."""
    subtotal   = sum(items)
    discounted = subtotal * (1 - discount_percent / 100)
    total      = discounted * (1 + tax_rate)
    return round(total, 2)

prices = [10.99, 5.49, 3.99, 8.75]

print("No discount:        ", calculate_total(prices))
print("10% discount:       ", calculate_total(prices, discount_percent=10))
print("20% off, no tax:    ", calculate_total(prices, discount_percent=20, tax_rate=0.0))
````

### ⚠️ The Mutable Default Argument Trap

This is one of the most common Python bugs — even for experienced developers.

**The problem:** Default values are evaluated **once** at function definition time.
If the default is a mutable object (`list`, `dict`, `set`), all calls share it.

**The fix:** Use `None` as the sentinel value and create the object inside the function.


````python
# ── BROKEN version — do NOT use this pattern ──────────────────────────────
def add_to_cart_broken(item, cart=[]):  # ❌ list created once!
    cart.append(item)
    return cart

print("Call 1:", add_to_cart_broken("Apples"))   # ['Apples']
print("Call 2:", add_to_cart_broken("Bread"))    # ['Apples', 'Bread']  ← Bug!
print("Call 3:", add_to_cart_broken("Milk"))     # ['Apples', 'Bread', 'Milk']  ← Still wrong!

# ── SAFE version — use None sentinel ───────────────────────────────────────
def add_to_cart(item, cart=None):  # ✅ correct
    if cart is None:
        cart = []       # fresh list every single call
    cart.append(item)
    return cart

print("\nSafe call 1:", add_to_cart("Apples"))
print("Safe call 2:", add_to_cart("Bread"))     # ['Bread']  ← Correct!
print("Safe call 3:", add_to_cart("Milk"))      # ['Milk']   ← Correct!
````

### 2.2 Variable Arguments: `*args` and `**kwargs`

| Syntax | Name | Receives |
|---|---|---|
| `*args` | Variable positional args | A **tuple** of extra positional arguments |
| `**kwargs` | Variable keyword args | A **dict** of extra keyword arguments |


````python
# ── *args — accept any number of positional arguments ──────────────────────
def total_score(*scores):
    """Accept any number of scores and return their sum."""
    print(f"  Received {len(scores)} scores: {scores}")  # scores is a tuple
    return sum(scores)

print("One score:  ", total_score(85))
print("Three scores:", total_score(85, 72, 91))
print("Five scores:", total_score(100, 90, 80, 70, 60))

# ── Required parameter before *args ────────────────────────────────────────
def announce_winners(event, *players):
    """Announce winners — event is required, players are variable."""
    print(f"Results for {event}:")
    for rank, player in enumerate(players, 1):
        print(f"  {rank}. {player}")

announce_winners("100m Sprint", "Alice", "Bob", "Charlie")
announce_winners("Long Jump", "Diana")
````

````python
# ── **kwargs — accept any number of keyword arguments ──────────────────────
def create_profile(name, student_id, **attributes):
    """Create a student profile from any number of extra attributes."""
    profile = {"name": name, "id": student_id}
    profile.update(attributes)   # attributes is a dict
    return profile

alice = create_profile("Alice", "S001", year=2, gpa=3.8, club="Chess")
bob   = create_profile("Bob",   "S002", year=1, scholarship=True)

print("Alice:", alice)
print("Bob:  ", bob)
````

````python
# ── Argument unpacking with * and ** at the call site ──────────────────────
def book_details(title, author, year):
    print(f"{title} by {author} ({year})")

# Unpack a list into positional arguments
book_info = ["Clean Code", "Robert Martin", 2008]
book_details(*book_info)           # ← same as book_details("Clean Code", "Robert Martin", 2008)

# Unpack a dict into keyword arguments
book_dict = {"title": "Fluent Python", "author": "Luciano Ramalho", "year": 2022}
book_details(**book_dict)
````

### 2.3 Specialized Sorting

Python's `sorted()` function and `list.sort()` both accept a `key` parameter —
a function applied to each item before comparing.

| Feature | `sorted(iterable)` | `list.sort()` |
|---|---|---|
| Modifies original | No — returns new list | Yes — in-place |
| Input | Any iterable | Lists only |
| Return value | New sorted list | `None` |


````python
# ── Sorting simple lists ────────────────────────────────────────────────────
scores = [72, 45, 91, 60, 88]

print("Ascending: ", sorted(scores))                # [45, 60, 72, 88, 91]
print("Descending:", sorted(scores, reverse=True))  # [91, 88, 72, 60, 45]
print("Original:  ", scores)                         # unchanged — sorted() is non-destructive

names = ["Charlie", "Alice", "Diana", "Bob"]
print("Names:     ", sorted(names))                 # ['Alice', 'Bob', 'Charlie', 'Diana']
print("By length: ", sorted(names, key=len))        # ['Bob', 'Alice', 'Diana', 'Charlie']
````

`key=len` tells `sorted()`:
- "Before comparing items, apply len() to each item and sort using that result."

````python
# ── Sorting a list of dicts with a lambda key ──────────────────────────────
students = [
    {"name": "Charlie", "score": 72, "age": 20},
    {"name": "Alice",   "score": 91, "age": 19},
    {"name": "Bob",     "score": 55, "age": 22},
    {"name": "Diana",   "score": 91, "age": 18},
]

# Sort by score descending
by_score = sorted(students, key=lambda s: s["score"], reverse=True)
print("By score (desc):")
for s in by_score:
    print(f"  {s['name']}: {s['score']}")

# Sort by score descending, then name ascending (tie-breaking)
by_score_name = sorted(students, key=lambda s: (-s["score"], s["name"]))
print("\nBy score desc, name asc:")
for s in by_score_name:
    print(f"  {s['name']}: {s['score']}")
````

When Python sorts tuples, it compares them left to right.

So:

- Compare the first element (-score)
- If they're equal, compare the second element (name)
- If those are equal, compare the third element (if there is one), and so on.

````python
from operator import itemgetter

students = [
    {"name": "Charlie", "score": 72},
    {"name": "Alice",   "score": 91},
    {"name": "Bob",     "score": 55},
]

# itemgetter — more readable than lambda for dict keys
sorted_by_name  = sorted(students, key=itemgetter("name"))
sorted_by_score = sorted(students, key=itemgetter("score"), reverse=True)

print("By name: ", [s["name"] for s in sorted_by_name])
print("By score:", [s["score"] for s in sorted_by_score])

# # ── lambda recap ────────────────────────────────────────────────────────────
# # lambda syntax: lambda arguments: expression
# square = lambda x: x**2     # same as def square(x): return x**2
# print("\nsquare(5):", square(5))

# items = [("Laptop", 899), ("Mouse", 25), ("Keyboard", 75)]
# cheapest = min(items, key=lambda item: item[1])
# print("Cheapest: ", cheapest)
````

### 3.1 `namedtuple` — Tuple with Named Fields

A `namedtuple` is a **tuple with named fields** — more readable and still immutable.

In Python, namedtuple is a way to create lightweight objects that behave like tuples but with named fields.

| | `namedtuple` | `dict` | Class |
|---|---|---|---|
| Field access | `obj.field` | `obj["field"]` | `obj.field` |
| Immutable | ✅ | ❌ | ❌ |
| Memory | Low | Higher | Medium |

Note: **Immutable** means that  cannot change it after it is created.

````python
from collections import namedtuple

# Define the named tuple type
Student = namedtuple("Student", ["name", "student_id", "score", "grade"])

# Create instances
alice   = Student("Alice",   "S001", 85, "B")
bob     = Student("Bob",     "S002", 42, "F")
charlie = Student("Charlie", "S003", 91, "A")

# Access fields by NAME — clear and readable
print(f"{alice.name}: {alice.score} ({alice.grade})")

# Still works as a regular tuple
print("Positional access:", alice[2])   # 85

# Sort a list of students by score
class_list = [alice, bob, charlie]
ranked = sorted(class_list, key=lambda s: s.score, reverse=True)
print("\nRanking:")
for rank, student in enumerate(ranked, 1):
    print(f"  {rank}. {student.name}: {student.score}")

# _asdict() — convert to dict
print("\nAs dict:", alice._asdict()) # returns a dict object

# _replace() — create new instance with changed fields (original unchanged)
updated = alice._replace(score=90, grade="A")
print("Updated:", updated)
print("Original:", alice)   # unchanged!
````

### 3.2 `deque` — Double-Ended Queue

A `deque` supports **O(1) operations at both ends** — `list` is O(n) for left-side operations.

| Operation | `list` | `deque` |
|---|---|---|
| Append to right | O(1) | O(1) |
| Append to left | **O(n) slow** | **O(1) fast** |
| Pop from left | **O(n) slow** | **O(1) fast** |


````python
from collections import deque

# ── Library computer queue ──────────────────────────────────────────────────
print("=== Library Computer Queue ===")
queue = deque()

queue.append("Alice")         # join at back
queue.append("Bob")
queue.append("Charlie")
print("Queue:", list(queue))  # ['Alice', 'Bob', 'Charlie']

queue.appendleft("Professor Smith")   # priority — join at front (O(1)!)
print("After priority:", list(queue))

next_user = queue.popleft()           # serve the next person
print(f"Serving: {next_user}")
print("Remaining:", list(queue))

# ── maxlen — fixed-size sliding window ─────────────────────────────────────
print("\n=== Recent Scores (Last 5) ===")
recent = deque(maxlen=5)

for score in [95, 80, 72, 88, 91, 78, 65, 99]:
    recent.append(score)
    print(f"  Added {score:3d}  →  Last 5: {list(recent)}")
````

````python
from collections import deque

d = deque([1, 2, 3, 4, 5])

d.rotate(2)
print(d)
# deque([4, 5, 1, 2, 3])

d.rotate(-2)
print(d)
# deque([1, 2, 3, 4, 5])
````

### 3.3 `ChainMap` — Layered Configuration

A `ChainMap` groups multiple dicts into a **single searchable view**
without copying any data. Keys are searched left to right — first match wins.

**Analogy:** Checking the local shop first, then the regional catalogue.


````python
from collections import ChainMap

# ── Configuration layers ────────────────────────────────────────────────────
local_settings  = {"theme": "dark",  "font_size": 14}
global_defaults = {"theme": "light", "font_size": 12, "language": "en", "timeout": 30}

config = ChainMap(local_settings, global_defaults)

print("theme:     ", config["theme"])      # 'dark'  — local wins
print("font_size: ", config["font_size"])  # 14      — local wins
print("language:  ", config["language"])   # 'en'    — from global
print("timeout:   ", config["timeout"])    # 30      — from global

# Updates go to the FIRST dict only
config["font_size"] = 16
print("\nAfter update — local:", local_settings)   # font_size changed
print("After update — global:", global_defaults)  # unchanged!
````

### 3.4 `Counter` — Count Anything

A `Counter` is a `dict` subclass for counting hashable objects.
Missing keys return `0` — no `KeyError`.


````python
from collections import Counter

# ── Count exam grades ────────────────────────────────────────────────────────
grades = ["A", "B", "A", "C", "B", "A", "F", "C", "B", "A"]
grade_count = Counter(grades)
print("Grade counts:", grade_count)

print("Count of A:  ", grade_count["A"])   # 4
print("Count of D:  ", grade_count["D"])   # 0 — no KeyError!
print("Top 2:       ", grade_count.most_common(2))

# ── Count words ──────────────────────────────────────────────────────────────
sentence = "the cat sat on the mat the cat sat"
word_count = Counter(sentence.split())
print("\nWord counts:", word_count)

# ── Counter arithmetic ───────────────────────────────────────────────────────
morning   = Counter({"Coffee": 5, "Tea": 3, "Water": 8})
afternoon = Counter({"Coffee": 7, "Tea": 6, "Juice": 4})

daily = morning + afternoon
diff  = afternoon - morning    # only positive differences kept

print("\nDaily totals:", daily)
print("Afternoon gains:", diff)
````

### 3.5 `OrderedDict` — Order-Aware Dictionary

Since Python 3.7, plain `dict` maintains insertion order.
`OrderedDict` still has two unique capabilities:

1. `move_to_end()` — move a key to the front or back
2. **Order-sensitive equality** — two `OrderedDict`s with the same items but different order are NOT equal


````python
from collections import OrderedDict

# ── LRU cache simulation ─────────────────────────────────────────────────────
cache = OrderedDict()
cache["page_1"] = "<html>Home</html>"
cache["page_2"] = "<html>About</html>"
cache["page_3"] = "<html>Contact</html>"

print("Initial order:", list(cache.keys()))

cache.move_to_end("page_1")              # mark as recently used → end
print("After access page_1:", list(cache.keys()))

cache.move_to_end("page_3", last=False)  # move to FRONT
print("After front page_3: ", list(cache.keys()))

# ── Order-sensitive equality ─────────────────────────────────────────────────
d1 = OrderedDict([("a", 1), ("b", 2)])
d2 = OrderedDict([("b", 2), ("a", 1)])
d3 = {"a": 1, "b": 2}
d4 = {"b": 2, "a": 1}

print("\nOrderedDict equality (same items, different order):", d1 == d2)   # False
print("Plain dict equality (same items, different order):  ", d3 == d4)    # True
````

### 3.6 `defaultdict` — Never a `KeyError`

A `defaultdict` **auto-creates missing keys** using a **factory function** you specify.

| Factory | Default value | Use case |
|---|---|---|
| `list` | `[]` | Grouping items |
| `int` | `0` | Counting |
| `set` | `set()` | Unique grouping |
| `lambda: value` | custom | Any default |


````python
from collections import defaultdict

# ── Grouping: subjects → list of enrolled students ──────────────────────────
enrolments = defaultdict(list)
data = [
    ("Alice",   "Maths"),
    ("Bob",     "Physics"),
    ("Alice",   "Physics"),
    ("Charlie", "Maths"),
    ("Bob",     "Maths"),
]

for student, subject in data:
    enrolments[subject].append(student)  # no if-check needed!

print("Enrolments:")
for subject, students in enrolments.items():
    print(f"  {subject}: {students}")

# ── Counting ─────────────────────────────────────────────────────────────────
votes = ["Alice", "Bob", "Alice", "Charlie", "Alice", "Bob"]
vote_count = defaultdict(int)
for candidate in votes:
    vote_count[candidate] += 1   # starts at 0 automatically

print("\nVote counts:", dict(vote_count))

# ── Custom default with lambda ────────────────────────────────────────────────
inventory = defaultdict(lambda: {"quantity": 0, "price": 0.0})
inventory["Laptop"]["quantity"] = 5
inventory["Laptop"]["price"] = 899.99
print("\nLaptop:", inventory["Laptop"])
print("Keyboard (auto-created):", inventory["Keyboard"])
````

### 3.7 `UserDict`, `UserList`, `UserString`

These wrapper classes make it safe to **subclass** the built-in containers.

**Why not subclass `dict` directly?**
Methods like `update()` may bypass your overridden `__setitem__()` — silently!
`UserDict` guarantees **all** writes go through your custom logic.


````python
from collections import UserDict

class GradeBook(UserDict):
    """A validated grade book — only accepts scores 0-100."""

    def __setitem__(self, student, grade):
        if not isinstance(grade, (int, float)):
            raise TypeError(f"Grade must be numeric, got {type(grade).__name__}.")
        if not (0 <= grade <= 100):
            raise ValueError(f"Grade {grade} is out of range (0-100).")
        super().__setitem__(student, grade)

    def class_average(self):
        if not self.data:
            return 0.0
        return sum(self.data.values()) / len(self.data)

    def passing_students(self, pass_mark=50):
        return [name for name, grade in self.data.items() if grade >= pass_mark]


book = GradeBook()
book["Alice"]   = 85
book["Bob"]     = 72
book["Charlie"] = 91

print("Grade book:     ", dict(book))
print("Class average:  ", round(book.class_average(), 1))
print("Passing:        ", book.passing_students())

# Validation
try:
    book["Diana"] = 105
except ValueError as e:
    print(f"ValueError: {e}")

try:
    book["Eve"] = "A+"
except TypeError as e:
    print(f"TypeError: {e}")
````

````python
from collections import UserList, UserString

# ── UserList — list with capacity and score validation ──────────────────────
class BoundedScoreList(UserList):
    """A list that only accepts valid scores and enforces max capacity."""
    def __init__(self, max_size=30, data=None):
        super().__init__(data or [])
        self.max_size = max_size

    def append(self, item):
        if len(self.data) >= self.max_size:
            raise OverflowError(f"Class is full (max {self.max_size}).")
        if not isinstance(item, (int, float)) or not (0 <= item <= 100):
            raise ValueError(f"Invalid score: {item}. Must be 0-100.")
        super().append(item)

    def average(self):
        return sum(self.data) / len(self.data) if self.data else 0.0


scores = BoundedScoreList(max_size=5)
for s in [85, 72, 91, 55]:
    scores.append(s)
print("Scores:", scores)
print("Average:", round(scores.average(), 1))

# ── UserString — string with format validation ──────────────────────────────
class StudentID(UserString):
    """Student IDs must start with S followed by digits."""
    def __init__(self, value):
        s = str(value)
        if not (s.startswith("S") or s.startswith("s")) or not s[1:].isdigit():
            raise ValueError(f"Invalid ID: {value!r}. Format: S + digits.")
        super().__init__(s.upper())

sid = StudentID("s042")
print("\nStudent ID:", sid)      # S042
print("Length:    ", len(sid))   # 4
print("Upper:     ", sid.upper())

try:
    bad = StudentID("ABC123")
except ValueError as e:
    print(f"ValueError: {e}")
````

### Mini Challenge 1 — Student Report Generator

**Goal:** Combine `namedtuple`, `Counter`, and sorted output to produce a class report.

Write a function `generate_report(students)` that accepts a list of
`Student` namedtuples and prints:
1. A ranked list (highest score first)
2. A grade distribution (A, B, C, D, F counts)
3. The class average


````python
from collections import namedtuple, Counter

Student = namedtuple("Student", ["name", "student_id", "score", "grade"])

class_data = [
    Student("Alice",   "S001", 85,  "B"),
    Student("Bob",     "S002", 42,  "F"),
    Student("Charlie", "S003", 91,  "A"),
    Student("Diana",   "S004", 67,  "C"),
    Student("Eve",     "S005", 78,  "B"),
    Student("Frank",   "S006", 55,  "D"),
    Student("Grace",   "S007", 93,  "A"),
    Student("Henry",   "S008", 72,  "C"),
]


def generate_report(students):
    """Generate a class report from a list of Student namedtuples."""
    print("=" * 40)
    print("         CLASS REPORT")
    print("=" * 40)

    # 1. Ranked list
    ranked = sorted(students, key=lambda s: s.score, reverse=True)
    print("\n📊 Rankings:")
    for rank, s in enumerate(ranked, 1):
        print(f"  {rank:2}. {s.name:<10} {s.score:3d}  Grade: {s.grade}")

    # 2. Grade distribution using Counter
    grade_dist = Counter(s.grade for s in students)
    print("\n📈 Grade Distribution:")
    for grade in ["A", "B", "C", "D", "F"]:
        count = grade_dist[grade]
        bar = "█" * count
        print(f"  {grade}: {bar} ({count})")

    # 3. Class average
    avg = sum(s.score for s in students) / len(students)
    print(f"\n✅ Class Average: {avg:.1f}")
    print("=" * 40)


generate_report(class_data)
````

### Mini Challenge 2 — Shopping Cart System

**Goal:** Use `UserDict` to build a validated shopping cart with
quantity tracking, total calculation, and a discount system.


````python
from collections import UserDict

class ShoppingCart(UserDict):
    """
    A validated shopping cart.
    Keys are product names (str), values are (price, quantity) tuples.
    """

    def __setitem__(self, product, value):
        if not isinstance(product, str) or not product:
            raise TypeError("Product name must be a non-empty string.")
        if not isinstance(value, tuple) or len(value) != 2:
            raise ValueError("Value must be a (price, quantity) tuple.")
        price, qty = value
        if price < 0 or qty < 0:
            raise ValueError("Price and quantity must be non-negative.")
        super().__setitem__(product, value)

    def total(self):
        """Return the cart total before discount."""
        return sum(price * qty for price, qty in self.data.values())

    def discounted_total(self, percent):
        """Return the total after applying a percentage discount."""
        return round(self.total() * (1 - percent / 100), 2)

    def receipt(self, discount_percent=0):
        """Print a formatted receipt."""
        print("=" * 40)
        print("         SHOPPING RECEIPT")
        print("=" * 40)
        for product, (price, qty) in self.data.items():
            print(f"  {product:<20} £{price:6.2f}  x{qty}")
        print("-" * 40)
        print(f"  Subtotal:          £{self.total():7.2f}")
        if discount_percent > 0:
            disc_amt = self.total() - self.discounted_total(discount_percent)
            print(f"  Discount ({discount_percent}%):     -£{disc_amt:6.2f}")
            print(f"  TOTAL:             £{self.discounted_total(discount_percent):7.2f}")
        else:
            print(f"  TOTAL:             £{self.total():7.2f}")
        print("=" * 40)


# Demo
cart = ShoppingCart()
cart["Laptop"]   = (899.99, 1)
cart["Notebook"] = (2.99,   3)
cart["Pen Pack"] = (4.50,   2)

cart.receipt(discount_percent=10)

# Validation test
try:
    cart[""] = (10, 1)          # empty product name
except TypeError as e:
    print(f"\nTypeError: {e}")

try:
    cart["USB"] = (15.99, -1)   # negative quantity
except ValueError as e:
    print(f"ValueError: {e}")
````

## Unit Summary

### Comprehensions

| Type | Syntax | Produces |
|---|---|---|
| List | `[expr for x in it if cond]` | `list` |
| Dict | `{k: v for x in it if cond}` | `dict` |
| Set | `{expr for x in it if cond}` | `set` |
| Generator | `(expr for x in it if cond)` | lazy iterator |

**Golden rule:** a comprehension should fit on one readable line.

---

### Functions

| Feature | Use when |
|---|---|
| Default params | Most callers want the same value |
| `*args` | Variable number of positional args |
| `**kwargs` | Variable number of keyword args |
| `lambda` | Short, one-expression key functions |
| `sorted(key=...)` | Sorting beyond alphabetical/numeric |

**Critical:** Never use `list`, `dict`, or `set` as a default parameter.
