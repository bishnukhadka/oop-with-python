## Part 1 — Classes in Python: Why OOP?

### 1.1 The Problem with Procedural Code

In Unit 1 you wrote procedural code — a top-to-bottom list of instructions.  
It works for small programs, but quickly becomes unmanageable.

````python
# ── PROCEDURAL style (Unit 1 approach) ──────────────────────────────────────
student_name  = "Alice"
student_score = 85
student_age   = 19

def get_grade(score):
    if score >= 90: return "A"
    elif score >= 75: return "B"
    elif score >= 60: return "C"
    else: return "F"

print(f"{student_name}: {get_grade(student_score)}")
# Output: Alice: B

# Problem: For 30 students you need 90 variables.
# Data and behaviour live in completely separate places. As programs grow, chaos!
````

### 1.2 The OOP Solution — Bundle Data + Behaviour

OOP wraps related data and behaviour into a single unit called a **class**.

| Concept | Analogy | Python |
|---------|---------|--------|
| **Class** | Cookie cutter / Blueprint | `class Student:` |
| **Object (Instance)** | One specific cookie | `alice = Student(...)` |
| **Attribute** | Cookie's topping / colour | `alice.score = 85` |
| **Method** | What the object can do | `alice.get_grade()` |

### 1.3 The Four Pillars of OOP

| Pillar | Definition | Example |
|--------|-----------|---------|
| **Encapsulation** | Bundle data + behaviour; hide internals | `BankAccount` hides balance; exposes `deposit()` |
| **Inheritance** | Child class reuses and extends parent | `SavingsAccount` IS A `BankAccount` |
| **Polymorphism** | Same call → different behaviour | `.attack()` on Warrior vs Mage vs Archer |
| **Abstraction** | Show only what is needed | `account.withdraw(100)` hides the maths |

````python
# ── OOP style — same result, far better structure ────────────────────────────
class Student:
    def __init__(self, name, score):    # constructor
        self.name  = name    # attribute: data stored on THIS student
        self.score = score   # each object gets its OWN copy of these

    def get_grade(self):     # method: behaviour belonging to a student
        if self.score >= 90: return "A"
        if self.score >= 75: return "B"
        if self.score >= 60: return "C"
        return "F"


# Create as many students as you like — one class handles them all
alice   = Student("Alice",   85)
bob     = Student("Bob",     72)
charlie = Student("Charlie", 91)

for student in [alice, bob, charlie]:
    print(f"{student.name}: {student.get_grade()}")
# Alice: B
# Bob: C
# Charlie: A
````

## Part 2 — Creating Classes and Objects

### 2.1 `__init__` and `self`

`__init__` is the **constructor** — called automatically when you create an object.  
`self` refers to **the specific object being created or used** — always the first parameter, never passed explicitly.

### 2.2 Instance Variables vs Class Variables

| | Instance Variable | Class Variable |
|--|-------------------|----------------|
| **Defined** | Inside `__init__` via `self.x = ...` | Class body, outside `__init__` |
| **Belongs to** | One specific object | ALL objects of the class |
| **Access** | `object.x` | `ClassName.x` (preferred) |
| **Example** | `self.owner`, `self.balance` | `BankAccount.bank_name` |

````python
class BankAccount:
    # CLASS VARIABLE — shared by ALL bank account objects
    bank_name     = "National Python Bank"
    account_count = 0

    def __init__(self, owner, account_number, balance=0):
        # INSTANCE VARIABLES — unique to each account object
        self.owner          = owner
        self.account_number = account_number
        self.balance        = balance
        self.transactions   = []   # empty transaction log

        BankAccount.account_count += 1   # update the class-level counter

    def deposit(self, amount):
        """Add funds to this account."""
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        self.transactions.append(f"+{amount:.2f}")
        print(f"Deposited Rs{amount:,.2f}. Balance: Rs{self.balance:,.2f}")

    def withdraw(self, amount):
        """Remove funds from this account."""
        if amount <= 0:
            raise ValueError("Amount must be positive.")
        if amount > self.balance:
            raise ValueError(f"Insufficient funds. Balance: Rs{self.balance:,.2f}")
        self.balance -= amount
        self.transactions.append(f"-{amount:.2f}")
        print(f"Withdrew Rs{amount:,.2f}. Balance: Rs{self.balance:,.2f}")

    def __str__(self):
        return (f"[{self.bank_name}] {self.owner}"
                f" ({self.account_number}) — Rs{self.balance:,.2f}")

# ── Instantiation — create three independent account objects ─────────────────
alice_acc   = BankAccount("Alice",   "ACC001", 5000)
bob_acc     = BankAccount("Bob",     "ACC002", 1200)
charlie_acc = BankAccount("Charlie", "ACC003")   # default balance = 0

print(alice_acc)     # [National Python Bank] Alice (ACC001) — Rs5,000.00
print(bob_acc)       # [National Python Bank] Bob (ACC002) — Rs1,200.00
print(f"Accounts opened: {BankAccount.account_count}")  # 3
````

> ⚠️ **Gotcha:** `alice_acc.bank_name = "New Bank"` creates an **instance variable** on  
> `alice_acc` only — it does NOT change the class variable for everyone!

````python
# Demonstrate the class variable gotcha
print("Before:", BankAccount.bank_name)    # National Python Bank
print("alice :", alice_acc.bank_name)      # National Python Bank

# This creates a SHADOW instance variable on alice only
alice_acc.bank_name = "Alice Private Bank"

print("After alice_acc.bank_name = 'Alice Private Bank':")
print("alice :", alice_acc.bank_name)      # Alice Private Bank  (instance var)
print("bob   :", bob_acc.bank_name)        # National Python Bank (class var unchanged)
print("class :", BankAccount.bank_name)    # National Python Bank (class var unchanged)

# Fix: delete the shadow to restore access to the class variable
del alice_acc.bank_name
print("After del alice_acc.bank_name:")
print("alice :", alice_acc.bank_name)      # National Python Bank (falls through to class var)
````

## Part 3 — Instance Methods and Special Methods (Dunders)

### 3.1 Instance Methods

Instance methods are functions defined inside a class.  
They receive `self` as the first argument and can read/modify the object's data.

````python
class Student:
    school_name = "Python Academy"

    def __init__(self, name, score, age):
        self.name  = name
        self.score = score
        self.age   = age

    def get_grade(self):
        if self.score >= 90: return "A"
        if self.score >= 75: return "B"
        if self.score >= 60: return "C"
        return "F"

    def apply_bonus(self, points):
        """Add bonus points; cap at 100."""
        self.score = min(self.score + points, 100)
        print(f"Bonus applied. {self.name}'s score: {self.score}")

    def is_passing(self):
        return self.score >= 60

    def describe(self):
        """Print a summary — calls other methods via self."""
        status = "PASS" if self.is_passing() else "FAIL"
        print(f"Student: {self.name:15} | Score: {self.score:3} "
              f"| Grade: {self.get_grade()} | {status}")


alice = Student("Alice", 85, 19)
bob   = Student("Bob",   55, 20)

alice.describe()           # Student: Alice           | Score:  85 | Grade: B | PASS
bob.describe()             # Student: Bob             | Score:  55 | Grade: F | FAIL
alice.apply_bonus(8)       # Bonus applied. Alice's score: 93
alice.describe()           # Student: Alice           | Score:  93 | Grade: A | PASS
````

### 3.2 Special (Dunder) Methods

Dunder methods let your objects work with Python's built-in syntax and functions.

| Dunder Method | Triggered by | Common Use |
|---------------|-------------|------------|
| `__str__` | `print(obj)`, f-strings | Human-readable description |
| `__repr__` | `repr(obj)`, REPL | Unambiguous developer string |
| `__len__` | `len(obj)` | Size of a collection |
| `__eq__` | `obj1 == obj2` | Value equality |
| `__lt__` | `obj1 < obj2`, `sort()` | Ordering |
| `__contains__` | `item in obj` | Membership |
| `__add__` | `obj1 + obj2` | Addition |

> Without `__str__`, `print(obj)` shows `<__main__.Student object at 0x7f...>` — not useful!

````python
class LibraryBook:
    """A library book — demonstrates all key dunder methods."""

    library_name = "City Central Library"

    def __init__(self, title, author, isbn, pages):
        self.title          = title
        self.author         = author
        self.isbn           = isbn
        self.pages          = pages
        self.is_checked_out = False
        print(f'{self.title} object instantiated.') # uncomment this for checking the instantiation

    def check_out(self):
        if self.is_checked_out:
            print(f"Sorry — '{self.title}' is already checked out.")
        else:
            self.is_checked_out = True
            print(f"'{self.title}' checked out successfully.")

    def return_book(self):
        self.is_checked_out = False
        print(f"'{self.title}' returned. Thank you!")

    def __str__(self):
        """Human-readable — used by print() and f-strings."""

        print(f'__str__() method called.')

        status = "Checked Out" if self.is_checked_out else "Available"
        return f'"{self.title}" by {self.author} [{status}]'

    def __repr__(self):
        """Developer string — used in the REPL and for debugging."""

        print(f'__repr__() method called.')

        return (f"LibraryBook(title={self.title!r}, "
                f"author={self.author!r}, isbn={self.isbn!r})")

    def __len__(self):
        """len(book) returns the page count."""
        return self.pages

    def __eq__(self, other):
        """Two books are equal if they share the same ISBN."""
        if not isinstance(other, LibraryBook):
            return NotImplemented
        return self.isbn == other.isbn


book1 = LibraryBook("The Python Handbook", "A. Smith",  "978-001", 420)
book2 = LibraryBook("Clean Code",          "R. Martin", "978-013", 431)
book3 = LibraryBook("The Python Handbook", "A. Smith",  "978-001", 420)  # same as book1

print()
print(book1)                         # "The Python Handbook" by A. Smith [Available]

print()
print(repr(book1))                   # LibraryBook(title='The Python Handbook', ...)

print()
print(f"Pages: {len(book1)}")        # Pages: 420

print()
book1.check_out()                    # 'The Python Handbook' checked out successfully.
book1.check_out()                    # Sorry — 'The Python Handbook' is already checked out.

print()
book1.return_book()                  # 'The Python Handbook' returned. Thank you!
print(f"book1 == book3 : {book1 == book3}")   # True  (same ISBN)
print(f"book1 == book2 : {book1 == book2}")   # False (different ISBN)
````

### 3.3 A Complete Class — The Shopping Cart

This example puts together everything: `__init__`, instance methods, and dunder methods.

````python
class ShoppingCart:
    """A shopping cart for an online store."""

    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.items         = {}   # {name: (price, quantity)}

    def add_item(self, name, price, quantity=1):
        """Add an item, or increase its quantity if already present."""
        if name in self.items:
            existing_price, existing_qty = self.items[name]
            self.items[name] = (existing_price, existing_qty + quantity)
        else:
            self.items[name] = (price, quantity)
        print(f"Added {quantity}x {name} @ Rs{price:.2f}")

    def remove_item(self, name):
        if name in self.items:
            del self.items[name]
            print(f"Removed '{name}' from cart.")
        else:
            print(f"'{name}' is not in the cart.")

    def get_total(self):
        return sum(price * qty for price, qty in self.items.values())

    def apply_discount(self, percent):
        discount = self.get_total() * (percent / 100)
        print(f"Discount: {percent}% = Rs{discount:.2f}")
        return self.get_total() - discount

    def __len__(self):
        return len(self.items)

    def __contains__(self, item_name):
        """method that defines how the in operator works for your custom object."""
        return item_name in self.items

    def __str__(self):
        if not self.items:
            return f"{self.customer_name}'s cart is empty."
        lines = [f"=== {self.customer_name}'s Cart ==="]
        for name, (price, qty) in self.items.items():
            subtotal = price * qty
            lines.append(
                f"  {name:<20} {qty:>2}x Rs{price:>7.2f} = Rs{subtotal:>8.2f}")
        lines.append("-" * 47)
        lines.append(f"  {'TOTAL':<33} Rs{self.get_total():>8.2f}")
        return "".join(lines)


cart = ShoppingCart("Alice")
cart.add_item("Python Textbook", 450.00, 1)
cart.add_item("USB Hub",         120.00, 2)
cart.add_item("Notebook",         30.00, 3)
cart.add_item("Python Textbook", 450.00, 1)   # adds a second copy

print()
print(cart)
print()
print(f"Items in cart : {len(cart)}")                # 3
print(f"Has USB Hub?  : {'USB Hub' in cart}")        # True
print(f"Has Mouse?    : {'Mouse' in cart}")           # False
print()
total_after_discount = cart.apply_discount(10)
print(f"Total after 10% off: Rs{total_after_discount:.2f}")
````

## Part 4 — Inheritance

### 4.1 The Is-A Relationship

Inheritance lets a **child class** reuse and extend a **parent class**.

```
"Is-A" signal:
  SavingsAccount  IS A  BankAccount    ✅ use inheritance
  GraduateStudent IS A  Student        ✅ use inheritance
  ShoppingCart    HAS   Items          ❌ use composition (not inheritance)
```

The child inherits ALL parent attributes and methods automatically, then  
adds new ones **or** overrides existing ones.

````python
# ── PARENT (Base) CLASS ───────────────────────────────────────────────────────
class BankAccount:
    bank_name = "National Python Bank"

    def __init__(self, owner, account_number, balance=0):
        self.owner          = owner
        self.account_number = account_number
        self.balance        = balance
        self.transactions   = []

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        self.transactions.append(f"+Rs{amount:.2f}")
        print(f"Deposited Rs{amount:,.2f}. Balance: Rs{self.balance:,.2f}")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be positive.")
        if amount > self.balance:
            raise ValueError(f"Insufficient. Balance: Rs{self.balance:,.2f}")
        self.balance -= amount
        self.transactions.append(f"-Rs{amount:.2f}")
        print(f"Withdrew Rs{amount:,.2f}. Balance: Rs{self.balance:,.2f}")

    def get_statement(self):
        print(f"Statement: {self.owner} ({self.account_number})")
        for t in self.transactions:
            print(f"  {t}")
        print(f"  Balance: Rs{self.balance:,.2f}")

    def __str__(self):
        return (f"[{self.bank_name}] {self.owner}"
                f" ({self.account_number}) — Rs{self.balance:,.2f}")


# ── CHILD CLASS 1: SavingsAccount — inherits and EXTENDS BankAccount ──────────
class SavingsAccount(BankAccount):
    def __init__(self, owner, account_number, balance=0, interest_rate=0.05):
        super().__init__(owner, account_number, balance)  # run parent's __init__
        self.interest_rate = interest_rate                # NEW attribute

    def apply_interest(self):
        """NEW method — unique to SavingsAccount."""
        interest = self.balance * self.interest_rate
        self.balance += interest
        self.transactions.append(f"INT +Rs{interest:.2f}")
        print(f"Interest +Rs{interest:,.2f}. Balance: Rs{self.balance:,.2f}")

    def __str__(self):
        base = super().__str__()   # reuse parent's __str__
        return f"{base} | Rate: {self.interest_rate*100:.1f}%"


# ── CHILD CLASS 2: LoanAccount — OVERRIDES withdraw ───────────────────────────
class LoanAccount(BankAccount):
    def __init__(self, owner, account_number, credit_limit, rate=0.12):
        super().__init__(owner, account_number, balance=0)
        self.credit_limit = credit_limit
        self.rate         = rate

    def withdraw(self, amount):
        """OVERRIDE: loan allows balance to go negative up to the credit limit."""
        if amount <= 0:
            raise ValueError("Amount must be positive.")
        if self.balance - amount < -self.credit_limit:
            raise ValueError(f"Exceeds credit limit Rs{self.credit_limit:,.2f}")
        self.balance -= amount
        self.transactions.append(f"LN -Rs{amount:.2f}")
        print(f"Loan disbursed Rs{amount:,.2f}. "
              f"Outstanding: Rs{abs(min(self.balance, 0)):,.2f}")

    def __str__(self):
        return f"{super().__str__()} | Limit: Rs{self.credit_limit:,.2f}"


# ── Demo ─────────────────────────────────────────────────────────────────────
print("=== Savings Account ===")
savings = SavingsAccount("Alice", "SAV001", 10000, interest_rate=0.06)
print(savings)
savings.deposit(5000)       # inherited from BankAccount
savings.apply_interest()    # unique to SavingsAccount
savings.get_statement()     # inherited

print("\n=== Loan Account ===")
loan = LoanAccount("Bob", "LOAN001", credit_limit=50000)
print(loan)
loan.withdraw(20000)        # overridden version

print("\n=== isinstance() checks ===")
print(f"savings IS A SavingsAccount : {isinstance(savings, SavingsAccount)}")  # True
print(f"savings IS A BankAccount    : {isinstance(savings, BankAccount)}")     # True
print(f"savings IS A LoanAccount    : {isinstance(savings, LoanAccount)}")     # False
````

### 4.2 `super()` — Calling the Parent Class

`super()` safely accesses the parent class from inside a child method.  
Always call `super().__init__(...)` **first** in a child's `__init__`.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def greet(self):
        print(f"Hi, I am {self.name}, aged {self.age}.")


class Student(Person):
    def __init__(self, name, age, student_id, gpa):
        super().__init__(name, age)   # sets self.name and self.age
        self.student_id = student_id  # student-specific attrs
        self.gpa        = gpa

    def greet(self):
        super().greet()               # calls Person.greet first
        print(f"My student ID is {self.student_id} and GPA is {self.gpa:.2f}.")


class GraduateStudent(Student):
    def __init__(self, name, age, student_id, gpa, thesis_title):
        super().__init__(name, age, student_id, gpa)   # calls Student.__init__
        self.thesis_title = thesis_title

    def greet(self):
        super().greet()    # calls Student.greet, which calls Person.greet
        print(f"My thesis: '{self.thesis_title}'")


print("=== Student ===")
s = Student("Alice", 20, "S001", 3.75)
s.greet()

print("=== GraduateStudent ===")
g = GraduateStudent("Bob", 24, "GS001", 3.9, "AI in Education")
g.greet()
````

### 4.3 Method Resolution Order (MRO)

When Python looks for a method it searches classes in the **MRO** — a linearized order.  
View it with `ClassName.__mro__`.

````python
# Simple MRO
class A:
    def hello(self): print("Hello from A")

class B(A):
    def hello(self): print("Hello from B")

class C(B):
    pass   # no hello() — inherits from nearest ancestor

obj = C()
obj.hello()   # Hello from B  (B searched before A)
print(C.__mro__)
# (<class 'C'>, <class 'B'>, <class 'A'>, <class 'object'>)

print()

# Multiple inheritance
class Flyable:
    def fly(self):
        print(f"{self.name} is flying!")

class Swimmable:
    def swim(self):
        print(f"{self.name} is swimming!")

class Duck(Flyable, Swimmable):
    def __init__(self, name):
        self.name = name
    def quack(self):
        print(f"{self.name}: Quack!")

donald = Duck("Donald")
donald.fly()    # Donald is flying!
donald.swim()   # Donald is swimming!
donald.quack()  # Donald: Quack!
print(Duck.__mro__)
````

### Diamond Structure

      A
     / \
    B   C
     \ /
      D

**Problem Case**

````python
class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")
        A.greet(self)

class C(A):
    def greet(self):
        print("Hello from C")
        A.greet(self)

class D(B, C):
    pass

d = D()
d.greet()
````

````python
class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")
        super().greet()

class C(A):
    def greet(self):
        print("Hello from C")
        super().greet()

class D(B, C):
    def greet(self):
        print("Hello from D")
        super().greet()

d = D()
d.greet()
````

````python
print(D.mro())
````

❗ Why C is NOT called

Because:

- You never call super().greet()
- You explicitly jump to A.greet(self)
- That bypasses the whole inheritance chain (including C)

So C is completely ignored.

## Part 5 — Polymorphism

**Polymorphism = same method name, different behaviour** depending on the object type.  
The caller (loop, function) does NOT need to know the type — it just calls the method.

````python
class GameCharacter:
    def __init__(self, name, health, level):
        self.name   = name
        self.health = health
        self.level  = level

    def attack(self):
        print(f"{self.name} attacks for 10 damage.")   # base version

    def special_ability(self):
        print(f"{self.name} has no special ability.")

    def status(self):
        print(f"  [{self.name}] HP: {self.health} | Level: {self.level}")


class Warrior(GameCharacter):
    def __init__(self, name, health, level, weapon):
        super().__init__(name, health, level)
        self.weapon = weapon

    def attack(self):
        damage = 25 * self.level
        print(f"[Warrior] {self.name} swings {self.weapon} for {damage} damage!")

    def special_ability(self):
        print(f"[Warrior] {self.name} uses SHIELD BASH — stuns the enemy!")


class Mage(GameCharacter):
    def __init__(self, name, health, level, mana):
        super().__init__(name, health, level)
        self.mana = mana

    def attack(self):
        damage    = 40 * self.level
        self.mana = max(self.mana - 10, 0)
        print(f"[Mage] {self.name} casts fireball for {damage} magic damage! "
              f"(Mana: {self.mana})")

    def special_ability(self):
        print(f"[Mage] {self.name} uses TIME WARP!")


class Archer(GameCharacter):
    def __init__(self, name, health, level, arrows):
        super().__init__(name, health, level)
        self.arrows = arrows

    def attack(self):
        if self.arrows > 0:
            damage       = 20 * self.level
            self.arrows -= 1
            print(f"[Archer] {self.name} fires for {damage} damage! "
                  f"({self.arrows} arrows left)")
        else:
            print(f"[Archer] {self.name} is out of arrows!")

    def special_ability(self):
        print(f"[Archer] {self.name} uses RAIN OF ARROWS!")


# ── POLYMORPHISM — same loop, different behaviour ─────────────────────────────
party = [
    Warrior("Thorin",  200, 5, "Battle Axe"),
    Mage("Gandalf",    120, 8, mana=100),
    Archer("Legolas",  150, 6, arrows=20),
]

print("=== ATTACK ROUND ===")
for character in party:
    character.attack()         # calls DIFFERENT attack() for each type

print("=== SPECIAL ABILITIES ===")
for character in party:
    character.special_ability()

print("=== STATUS (inherited — not overridden) ===")
for character in party:
    character.status()
````

### 5.2 Duck Typing

> *"If it walks like a duck and quacks like a duck, it's a duck."*

Python does not care about the **type** of an object — only whether it has the **method** called.

````python
# Three completely unrelated classes — no shared parent
class PDFReport:
    def generate(self):
        return "report.pdf"

class ExcelReport:
    def generate(self):
        return "report.xlsx"

class HTMLReport:
    def generate(self):
        return "report.html"


# Works with ANY object that has generate() — duck typing
def run_report(reporter):
    filename = reporter.generate()
    print(f"Report saved: {filename}")


for reporter in [PDFReport(), ExcelReport(), HTMLReport()]:
    run_report(reporter)   # same call, different result
````

### 5.3 Operator Overloading with `@total_ordering`

Define comparison dunder methods to make `sort()`, `max()`, `min()` work with your objects.

````python
from functools import total_ordering

@total_ordering   # auto-generates __le__, __gt__, __ge__ from __eq__ + __lt__
class Student:
    def __init__(self, name, score):
        self.name  = name
        self.score = score

    def get_grade(self):
        for threshold, grade in [(90,"A"),(75,"B"),(60,"C")]:
            if self.score >= threshold:
                return grade
        return "F"

    def __str__(self):  return f"{self.name} ({self.score} — {self.get_grade()})"
    def __eq__(self, other): return self.score == other.score
    def __lt__(self, other): return self.score < other.score   # enables sort/min/max

    def __add__(self, other):
        """Combine two students into a new average-score student."""
        avg = (self.score + other.score) // 2
        return Student(f"{self.name}&{other.name}", avg)


students = [
    Student("Alice",   85),
    Student("Bob",     72),
    Student("Charlie", 91),
    Student("Diana",   68),
]

students.sort()   # uses __lt__
print("Sorted by score (lowest first):")
for s in students:
    print(f"  {s}")

print(f"Top student  : {max(students)}")
print(f"Lowest score : {min(students)}")

combo = students[0] + students[1]
print(f"Combined: {combo}")
````

### 6.2 Creating Custom Exception Classes

Create a **base exception** for your module, then specific subclasses.  
This lets callers catch all your errors broadly OR handle specific ones precisely.

````python
# ── Custom exception hierarchy for a bank system ──────────────────────────────
class BankError(Exception):
    """Base exception for all bank-related errors."""
    pass

class InsufficientFundsError(BankError):
    """Raised when a withdrawal exceeds the available balance."""
    def __init__(self, requested, available):
        self.requested = requested
        self.available = available
        super().__init__(
            f"Requested Rs{requested:,.2f} but only Rs{available:,.2f} available."
        )

class InvalidAmountError(BankError):
    """Raised when a transaction amount is zero or negative."""
    def __init__(self, amount):
        self.amount = amount
        super().__init__(f"Amount must be positive. Got: Rs{amount:,.2f}")

class AccountFrozenError(BankError):
    """Raised when a transaction is attempted on a frozen account."""
    def __init__(self, account_number):
        self.account_number = account_number
        super().__init__(f"Account '{account_number}' is frozen — contact the bank.")


# ── BankAccount that raises these custom errors ─────────────────────────────
class BankAccount:
    def __init__(self, owner, account_number, balance=0):
        self.owner          = owner
        self.account_number = account_number
        self.balance        = balance
        self.is_frozen      = False

    def deposit(self, amount):
        if self.is_frozen:  raise AccountFrozenError(self.account_number)
        if amount <= 0:     raise InvalidAmountError(amount)
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if self.is_frozen:       raise AccountFrozenError(self.account_number)
        if amount <= 0:          raise InvalidAmountError(amount)
        if amount > self.balance: raise InsufficientFundsError(amount, self.balance)
        self.balance -= amount
        return self.balance

    def freeze(self):
        self.is_frozen = True
        print(f"Account {self.account_number} frozen.")

    def __str__(self):
        state = "FROZEN" if self.is_frozen else "ACTIVE"
        return f"{self.owner} ({self.account_number}) [{state}] Rs{self.balance:,.2f}"


# ── Tests ────────────────────────────────────────────────────────────────────
acct = BankAccount("Alice", "ACC001", 500)
print(acct)

print("--- Test 1: valid withdrawal ---")
try:
    acct.withdraw(200)
    print(f"Balance: Rs{acct.balance:,.2f}")
except BankError as e:
    print(f"Error: {e}")

print("--- Test 2: insufficient funds ---")
try:
    acct.withdraw(1000)
except InsufficientFundsError as e:
    print(f"Insufficient! Need Rs{e.requested:,.2f}, have Rs{e.available:,.2f}")

print("--- Test 3: frozen account ---")
acct.freeze()
try:
    acct.deposit(100)
except AccountFrozenError as e:
    print(f"Frozen: {e}")

print("--- Test 4: invalid amount ---")
acct2 = BankAccount("Bob", "ACC002", 1000)
try:
    acct2.withdraw(-50)
except InvalidAmountError as e:
    print(f"Invalid: {e}")
````

### 6.3 `try` / `except` / `else` / `finally`

| Clause | When it runs |
|--------|-------------|
| `try` | Code that might raise an exception |
| `except` | If an exception occurred — handle it |
| `else` | Only if NO exception occurred |
| `finally` | ALWAYS — use for cleanup (close files, DB connections) |

````python
def safe_transaction(account, action, amount):
    """Safely attempt a deposit or withdrawal with full error handling."""
    print(f"Attempting {action} of Rs{amount:,.2f} for {account.owner}...")
    try:
        if action == "deposit":
            new_balance = account.deposit(amount)
        elif action == "withdraw":
            new_balance = account.withdraw(amount)
        else:
            raise ValueError(f"Unknown action: {action}")

    except AccountFrozenError as e:
        print(f"  FROZEN: {e}")
        new_balance = None

    except InsufficientFundsError as e:
        print(f"  INSUFFICIENT: need Rs{e.requested:,.2f}, "
              f"have Rs{e.available:,.2f}")
        new_balance = None

    except InvalidAmountError as e:
        print(f"  INVALID AMOUNT: {e}")
        new_balance = None

    except BankError as e:
        print(f"  BANK ERROR: {e}")
        new_balance = None

    else:
        # Only runs if no exception occurred
        print(f"  Success! New balance: Rs{new_balance:,.2f}")

    finally:
        # ALWAYS runs — log the attempt
        print(f"  Transaction logged.")

    return new_balance


test_account = BankAccount("Charlie", "ACC003", 800)

safe_transaction(test_account, "withdraw", 300)   # success
safe_transaction(test_account, "withdraw", 700)   # insufficient
safe_transaction(test_account, "deposit", -100)   # invalid
test_account.freeze()
safe_transaction(test_account, "deposit", 200)    # frozen
````

### 7.2 Building a Custom Iterator

Implement `__iter__` and `__next__` to control how an object is looped over.

````python
class CountDown:
    """An iterator that counts down from n to 1."""

    def __init__(self, start):
        self.current = start

    def __iter__(self):
        """Return self — this class IS the iterator."""
        return self

    def __next__(self):
        """Return the next value, or raise StopIteration when done."""
        if self.current <= 0:
            raise StopIteration
        value        = self.current
        self.current -= 1
        return value


print("Countdown:")
for n in CountDown(5):
    print(n, end=" ")   # 5 4 3 2 1
print()

# Manual iteration
cd = CountDown(3)
print(next(cd))   # 3
print(next(cd))   # 2
print(next(cd))   # 1
````

````python
# ── A practical iterator: StudentRoster ──────────────────────────────────────
class StudentRoster:
    """An iterable class roster."""

    def __init__(self, course_name):
        self.course_name = course_name
        self._students   = []

    def enroll(self, name, score):
        self._students.append((name, score))

    def __iter__(self):
        return iter(self._students)   # delegate to the list's iterator

    def __len__(self):
        return len(self._students)

    def __str__(self):
        return f"{self.course_name} ({len(self)} students)"


roster = StudentRoster("BT151CO — Python")
roster.enroll("Alice",   85)
roster.enroll("Bob",     72)
roster.enroll("Charlie", 91)

print(roster)
for name, score in roster:         # uses __iter__
    print(f"  {name}: {score}")

# Can loop multiple times — each loop gets a fresh iterator
print("Passing students (score >= 75):")
for name, score in roster:
    if score >= 75:
        print(f"  {name}")
````

### 7.3 Generator Functions — `yield`

A **generator function** uses `yield` to produce values **one at a time** (lazily).  
Python pauses execution at each `yield` and resumes on the next `next()` call.

````python
# Simple generator
def countdown_gen(n):
    print(f"[generator] Starting from {n}")
    while n > 0:
        yield n     # pause, return n, resume on next next()
        n -= 1
    print("[generator] Done!")

print("Creating the generator (nothing runs yet):")
gen = countdown_gen(3)

print("Calling next() manually:")
print(next(gen))   # [generator] Starting... yields 3
print(next(gen))   # yields 2
print(next(gen))   # yields 1

print("Using in a for loop:")
for n in countdown_gen(5):
    print(n, end=" ")
print()

# ── Memory: generator vs list ─────────────────────────────────────────────────
import sys

n = 100_000
list_sq = [x**2 for x in range(n)]   # ALL values built NOW
gen_sq  = (x**2 for x in range(n))   # ONE value at a time

print(f"List memory   : {sys.getsizeof(list_sq):>10,} bytes")
print(f"Generator mem : {sys.getsizeof(gen_sq):>10,} bytes")
print("First 5 (list)     :", list_sq[:5])
print("First 5 (generator):", [next(gen_sq) for _ in range(5)])
````

````python
# Practical generator: process a score pipeline
def score_pipeline(scores, passing=60):
    """Generator pipeline: skip invalid, cap at 100, yield only passing scores."""
    for score in scores:
        if score < 0:
            print(f"  [skip] invalid: {score}")
            continue
        score = min(score, 100)   # cap at 100
        if score >= passing:
            yield score


raw = [85, -5, 102, 45, 91, 0, 72, -10, 88, 50]
print("Processing scores:")
passing = list(score_pipeline(raw))
print(f"Passing scores : {passing}")
print(f"Class average  : {sum(passing) / len(passing):.1f}")


# yield from — delegate to another generator
def semester_one():
    yield from ["Alice", "Bob", "Charlie"]

def semester_two():
    yield from ["Diana", "Eve", "Frank"]

def all_students():
    yield from semester_one()
    yield from semester_two()

print("All students:")
for name in all_students():
    print(f"  {name}")
````

### 8.2 Scope — The LEGB Rule

Python searches for a name in this exact order:  
**L**ocal → **E**nclosing → **G**lobal → **B**uilt-in

| Scope | Where | Example |
|-------|-------|---------|
| **L** — Local | Inside the current function | `score = 85` inside `get_grade()` |
| **E** — Enclosing | Surrounding function (closures) | `count` in `make_counter()` |
| **G** — Global | Module (file) level | `PASSING = 60` at top of file |
| **B** — Built-in | Python's built-ins | `print`, `len`, `range` |

````python
# ── LEGB demonstration ────────────────────────────────────────────────────────
PASSING_SCORE = 60          # G — Global

def calculate_result(score):
    grade_label = "Grade"   # L — Local (not visible outside this function)
    if score >= PASSING_SCORE:          # reads Global
        return f"{grade_label}: PASS"   # reads Local
    return f"{grade_label}: FAIL"

print(calculate_result(75))   # Grade: PASS
print(calculate_result(40))   # Grade: FAIL
# print(grade_label)          # NameError! Local not accessible here


# ── Enclosing scope — closures ────────────────────────────────────────────────
def make_counter(start=0):
    count = start   # E — Enclosing

    def increment():
        nonlocal count   # tell Python: use the enclosing 'count'
        count += 1
        return count

    def reset():
        nonlocal count
        count = start
        print(f"Counter reset to {start}")

    return increment, reset

inc, rst = make_counter(0)
print(inc())   # 1
print(inc())   # 2
print(inc())   # 3
rst()          # Counter reset to 0
print(inc())   # 1


# ── global keyword (avoid — prefer return values) ─────────────────────────────
total_students = 0

def register_student(name):
    global total_students
    total_students += 1
    print(f"Registered {name}. Total: {total_students}")

register_student("Alice")
register_student("Bob")
print(f"Final count: {total_students}")
````

## Part 9 — Summary and Key Takeaways

### The Four Pillars — Quick Reference

| Pillar | Key Python Construct | Example |
|--------|---------------------|---------|
| **Encapsulation** | `class`, `self`, `__init__` | `BankAccount` hides balance |
| **Inheritance** | `class Child(Parent)`, `super()` | `SavingsAccount(BankAccount)` |
| **Polymorphism** | Method overriding, duck typing | `character.attack()` |
| **Abstraction** | Methods, dunder methods | `cart.get_total()` |

### Common Beginner Mistakes

| # | Mistake | Fix |
|---|---------|-----|
| 1 | Forgetting `self` as the first parameter | Always write `def method(self, ...)` |
| 2 | Forgetting `super().__init__()` in child | Always call `super()` first |
| 3 | Modifying class var via instance | Use `ClassName.var = ...` |
| 4 | Catching bare `except:` | Use `except Exception:` or specific type |
| 5 | Raising `BaseException` | Raise from `Exception` subclass |
| 6 | No `__str__` defined | Always add `__str__` for readable output |
| 7 | Calling `next()` on a list | Lists are iterable but not iterators — use `iter()` first |

````python
# ── Live recap — all concepts in one program ──────────────────────────────────
from functools import total_ordering

class CourseError(Exception):
    pass

class EnrollmentFullError(CourseError):
    def __init__(self, course, capacity):
        super().__init__(f"'{course}' is full (max {capacity} students).")

@total_ordering
class Student:
    school = "Python Academy"

    def __init__(self, name, gpa):
        self.name = name
        self.gpa  = gpa

    def __str__(self):  return f"{self.name} (GPA: {self.gpa:.2f})"
    def __repr__(self): return f"Student({self.name!r}, {self.gpa})"
    def __eq__(self, o): return self.gpa == o.gpa
    def __lt__(self, o): return self.gpa < o.gpa


class Course:
    def __init__(self, code, name, capacity):
        self.code     = code
        self.name     = name
        self.capacity = capacity
        self._roster  = []

    def enroll(self, student):
        if len(self._roster) >= self.capacity:
            raise EnrollmentFullError(self.name, self.capacity)
        self._roster.append(student)
        print(f"  {student.name} enrolled in {self.name}")

    def top_students(self, n=3):
        """Generator: yield top-n students by GPA."""
        for s in sorted(self._roster, reverse=True)[:n]:
            yield s

    def __len__(self):  return len(self._roster)
    def __iter__(self): return iter(self._roster)
    def __str__(self):
        return f"{self.code}: {self.name} ({len(self)}/{self.capacity} enrolled)"


cs101 = Course("CS101", "Introduction to Python", capacity=4)
students = [
    Student("Alice",   3.9),
    Student("Bob",     3.2),
    Student("Charlie", 3.7),
    Student("Diana",   3.5),
    Student("Eve",     3.8),   # will fail — course full
]

print("=== Enrolling ===")
for s in students:
    try:
        cs101.enroll(s)
    except EnrollmentFullError as e:
        print(f"  FULL: {e}")

print(f"\n{cs101}")

print("=== Top 3 by GPA ===")
for s in cs101.top_students(3):
    print(f"  {s}")

print("=== All enrolled (sorted) ===")
for s in sorted(cs101):
    print(f"  {s}")
````

````python
# ── BUGGY GENERATOR ──────────────────────────────────────────────────────────
def even_numbers(limit):
    """Generator: yield even numbers from 0 to limit (inclusive)."""
    n = 0
    while n < limit:     # BUG 1: should be <= to include limit
        if n % 2 = 0:    # BUG 2: = should be ==
            yield n
        n += 1

print(list(even_numbers(10)))   # expected: [0, 2, 4, 6, 8, 10]
````
